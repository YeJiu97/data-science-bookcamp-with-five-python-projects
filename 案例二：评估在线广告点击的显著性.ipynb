## 使用SciPy探索数据和概率之间的关系

SciPy是Scientific Python的缩写，为科学分析提供了许多有用的方法。

SciPy库包含了一个用来解决概率和统计问题的模块：scipy.stats。

````python
from scipy import stats
````

我们可以使用这个模块提供的函数之间实现第一个例子中：投掷一枚均质的硬币20次但是16次正面朝上的概率。

````python
num_heads = 16
num_flips = 20
prob_head = 0.5

prob = stats.binom_test(num_heads, num_flips, prob_head)
print(f"Probability of observing more than 15 heads or 15 tails is {prob:.17f}")
````

标准Python处理低值小数点的时候会返回17位置的有效数字，但是SciPy则是默认18位，所以这里进行了手工调整。

要处理这个弃用警告,需要使用 SciPy 库中的新函数 binomtest 来替换旧的 binom_test 函数。在新版本的 SciPy 中,binom_test 函数已被弃用,并将在未来的版本中被完全移除。

````python
import scipy.stats as stats

num_heads = 16
num_flips = 20 
prob_head = 0.5

result = stats.binomtest(num_heads, num_flips, prob_head, alternative='two-sided')
prob = result.pvalue

print(f"Probability of observing more than 16 heads or 15 tails is {prob:.17f}")
````

`alternative='two-sided'` 是 `scipy.stats.binomtest` 函数中的一个参数, 用于指定进行双侧检验还是单侧检验。

在二项分布检验中:

- 单侧检验 (single-sided) 是计算观测值大于或小于特定值的概率。
- 双侧检验 (two-sided) 是计算观测值大于或小于特定值的概率的总和。

具体来说:

- 'two-sided' 表示双侧检验,计算观测值偏离期望值的双侧概率
- 'less' 表示单侧检验,计算观测值小于等于给定值的概率
- 'greater' 表示单侧检验,计算观测值大于等于给定值的概率

在我们的例子中, `num_heads=16, num_flips=20`, 我们想计算观测到至少16次正面(大于15次)或至多4次正面(小于等于4次)的概率之和。这相当于计算偏离均值(正面10次)的双侧概率, 因此使用 `alternative='two-sided'`。

所以 `alternative='two-sided'` 表示我们进行的是双侧二项分布检验, 计算观测值偏离均值(成功概率为0.5时)的双侧概率。

````python
import scipy.stats as stats

num_heads = 16
num_flips = 20 
prob_head = 0.5

result = stats.binomtest(num_heads, num_flips, prob_head, alternative='less')
prob = result.pvalue

print(f"Probability of observing more than 15 heads or 15 tails is {prob:.17f}")
````

在这段代码中，alternative='less'表示计算观测到小于等于16次正面(大于等于4次反面)的概率。该概率值为0.99871158599853516，即99.87%。由于硬币是均匀的，因此观测到至多4次反面的概率等同于观测到至少16次正面的概率。


````python
import scipy.stats as stats

num_heads = 16
num_flips = 20 
prob_head = 0.5

result = stats.binomtest(num_heads, num_flips, prob_head, alternative='greater')
prob = result.pvalue

print(f"Probability of observing more than 15 heads or 15 tails is {prob:.17f}")
````

在这段代码中，alternative='greater'表示计算观测到大于等于16次正面(小于等于4次反面)的概率。该概率值为0.00590896606445312，即0.59%。由于硬币是均匀的，因此观测到至少16次正面的概率等同于观测到至多4次反面的概率。


总结为：

- alternative='greater'计算的是观测值严格大于给定值的概率累积。也就是说,不包括观测值等于给定值的情况。
- alternative='less'计算的是观测值严格小于给定值的概率累积。同样也不包括观测值等于给定值的情况。
- 而alternative='two-sided'计算的则包括了观测值等于、大于和小于给定值的所有情况的概率之和。

因此,在二项分布检验中:

- alternative='greater'给出的是P(X > k)的概率值
- alternative='less'给出的是P(X < k)的概率值
- alternative='two-sided'给出的是P(X >= k) + P(X <= n-k)的概率值之和

它们反映了不同的备择假设,用于根据观测数据检验原假设(如正面概率是否等于0.5)是否成立。您对其含义的理解是正确的。

````python
import scipy.stats as stats

n = 20  # 总次数
k = 16  # 成功次数
p = 0.5 # 成功概率

# 计算观测到至少 k+1 次成功的概率
res_greater = stats.binomtest(k+1, n, p, alternative='greater')
p_atleast_k_plus_1 = res_greater.pvalue

# 计算观测到至少 k 次成功的概率 
res_greater = stats.binomtest(k, n, p, alternative='greater')  
p_atleast_k = res_greater.pvalue

# 观测到恰好 k 次成功的概率 = 观测到至少 k 次成功的概率 - 观测到至少 k+1 次成功的概率
p_exactly_k = p_atleast_k - p_atleast_k_plus_1

print(f"Probability of observing exactly {k} heads out of {n} flips is: {p_exactly_k:.17f}")
````

````python
import scipy.stats as stats

num_heads = 16
num_flips = 20
prob_head = 0.5

# 使用stats.binom.pmf计算恰好4次和16次正面朝上的概率
probabilities = stats.binom.pmf([4, 16], num_flips, prob_head)

# 使用一个小的容忍值来比较两个浮点数是否“相等”
tolerance = 1e-9
assert abs(probabilities[0] - probabilities[1]) < tolerance, "The probabilities are not approximately equal"

print(f"Probability of observing exactly 4 heads or exactly 16 heads is {probabilities[0]:.17f}")
````

## 将均值作为中心性的度量

如果希望研究当地夏季第一周的温度，那么：

````python
import numpy as np
measurements = np.array([80, 77, 73, 61, 74, 79, 81])
````

````python
measurements.sort()
number_of_days = measurements.size
plt.plot(range(number_of_days), measurements)
plt.scatter(range(number_of_days), measurements)
plt.ylabel('Temperature')
plt.show()
````

简单的观察，可以认为中心温度在60华氏度到80华氏度之间。因此，可以简单的将中心值估计为大约70华氏度。我们将估计值量化为途中最低值和最高值的中点。

````python
# 查找中心温度值
difference = measurements.max() - measurements.min()
midpoint = measurements.min() + difference / 2

assert midpoint == (measurements.max() + measurements.min()) / 2
print(f"The midpoint temperature is {midpoint} defrees")
````

通过计算，中心温度为71华氏度。

通过可视化来展示：

````python
plt.plot(range(number_of_days), measurements)
plt.scatter(range(number_of_days), measurements)
plt.axhline(midpoint, color = 'k', linestyle = '--')
plt.ylabel("Temperature")
plt.show()
````

7个测量值中有6个高于中点。

````python
median = measurements[3]

print(f"The median temperature is {median} degrees")

plt.figure(figsize=(8, 4))  # 设置图像大小

plt.plot(range(number_of_days), measurements)
plt.scatter(range(number_of_days), measurements)

plt.axhline(median, color='g', linestyle='-.', label='median')

# 修正拼写错误
plt.axhline(midpoint, color='k', linestyle='--', label='midpoint') 

plt.legend()
plt.xlabel('Day')  # 添加 x 轴标签
plt.ylabel('Temperature')

plt.show()
````

使用与最小值的平方距离来惩罚中心值：

````python
def squared_distance(value1, value2):
    return (value1 - value2) ** 2

possible_centers = range(measurements.min(), measurements.max() + 1)
penalties = [squared_distance(center, 61) for center in possible_centers]
plt.plot(possible_centers, penalties)
plt.scatter(possible_centers, penalties)
plt.xlabel('Possible Centers')
plt.ylabel('Penalty')
plt.show()
````

随着中心值向着61的方向移动，惩罚在降低，但是它们与其他6个测量值的距离在增加。值或者对中心值进行排序。值或者对中心值进行排序。

因此，应该根据每个潜在的中心值到所有7个测量值的平方距离来惩罚它们。

````python
def sum_of_squared_distances(value, measurements):
    return sum(squared_distance(value, m) for m in measurements)

penalties = [sum_of_squared_distances(center, measurements)
            for center in possible_centers]

plt.plot(possible_centers, penalties)
plt.scatter(possible_centers, penalties)
plt.xlabel('Possible Centers')
plt.ylabel('Penalty')
plt.show()
````

从图中可以看出，在75华氏处可以得到最低惩罚。这个温度我们可以称之为“惩罚最小的中心值”。

计算了一个数字（给定值）与数组中的每个数字之间的差的平方，然后将所有这些差的平方值相加，得到一个总和。这种方法可以用来衡量给定值与数组中所有值之间的总平方距离，提供了一种方式来评估给定值与数据集的拟合程度或相似性。

````python
least_penalized = 75
assert least_penalized == possible_centers[np.argmin(penalties)]
````

`possible_centers[np.argmin(penalties)]` 是 Python 中的一种索引操作，它用于找到 `penalties` 列表中的最小值，并返回对应的 `possible_centers` 中的值。

让我们来解释一下这个表达式：

- `np.argmin(penalties)`：这个函数是 NumPy 库中的函数，它返回数组中最小值的索引位置。换句话说，它会找到 `penalties` 列表中最小值所在的位置。

- `possible_centers[]`：这是一个索引操作，用于从 `possible_centers` 中选择一个特定的值。在方括号中放入索引值，即可获取该索引位置上的值。

因此，`possible_centers[np.argmin(penalties)]` 这个表达式的含义是：找到 `penalties` 列表中的最小值的索引，然后使用这个索引从 `possible_centers` 中获取对应的值。这样就可以得到与最小惩罚值对应的中心值。

````python
plt.plot(range(number_of_days), measurements)
plt.scatter(range(number_of_days), measurements)
plt.axhline(midpoint, color = 'k', linestyle = '--', label = 'midpoint')
plt.axhline(median, color = 'g', linestyle = '-.', label = 'median')
plt.axhline(least_penalized, color = 'r', linestyle = '-',
            label = 'least penalized center')
plt.legend()
plt.ylabel('Temperature')
plt.show()
````

惩罚最少的中心值将测量到的温度相对均匀的分隔开：4个测量值在上面，3个测量值在下面，相对于中位数温度也更加的接近最冷温度记录的距离。

惩罚最少的中心值在衡量中心性时是一个很好的方法，因为它表示该中心值与数据集中的大多数数据点之间的距离较小，即该中心值更能够代表整个数据集的趋势和分布。这可以解释为什么惩罚最少的中心值被认为是中心性的一个很好的指标：

1. **数据代表性**: 惩罚最少的中心值意味着它与数据集中的大多数数据点距离较近。这表明该中心值对数据集的代表性较高，因为它更能够捕获数据集的整体特征和趋势。

2. **最小化总体误差**: 惩罚值的最小化通常意味着最小化了中心值与数据集中所有数据点之间的总体误差。因此，惩罚最小的中心值可以被视为使总体误差最小化的选择，这与我们希望中心值能够尽可能地代表数据集的整体特征一致。

3. **可解释性**: 最小化惩罚值的中心值通常更容易解释。因为它们更接近于数据集中的大多数数据点，所以它们的值更能够用来描述数据的中心趋势和集中程度。

综上所述，惩罚最少的中心值可以被认为是衡量中心性的一个很好的方法，因为它能够提供对数据集整体特征的更好代表，并且有助于最小化中心值与数据点之间的总体误差。

np.mean是计算数组元素的算术平均值,即将所有元素的总和除以元素个数。这是计算平均值最常见的方式。

np.average则提供了更多选项,可以指定在计算平均值时使用不同的权重。它还可以处理包含NaN(不是一个数字)值的数组,通过调整计算方式来避免NaN值对结果的影响。

以下是一个Python代码示例,展示了这两个函数的用法和区别:


````python
import numpy as np

# 创建一个数组
arr = np.array([1, 2, 3, 4, 5])

# 使用 np.mean() 计算平均值
mean_result = np.mean(arr)
print("Mean:", mean_result)  # 输出平均值

# 使用 np.average() 计算加权平均值
weights = np.array([0.1, 0.2, 0.3, 0.2, 0.2])  # 指定权重数组
weighted_mean_result = np.average(arr, weights=weights)
print("Weighted Mean:", weighted_mean_result)  # 输出加权平均值
````

当数组中的元素的权重是相等的时候，两者并没有区别，但是可以通过给np.average()权重来调整输出。

````python
equal_weights = [1] * 7
assert mean == np.average(measurements, weights = equal_weights)
````

````python
unequal_weights = [100] + [1] * 6
assert mean != np.average(measurements, weights = unequal_weights)
````

````python
weighted_mean = np.average([75, 77], weights = [9, 1])
print(f"The mean is {weighted_mean}")
assert weighted_mean == np.mean(9 * [75] + [77]) == weighted_mean
````

## 将方差作为离散型的度量

在概率论和统计学中，离散型随机变量是指其可能取值为有限个或可数个的随机变量。换句话说，离散型随机变量的取值集合是有限的或可数的，而不是连续的。

举例来说，投掷一枚骰子的结果就是一个离散型随机变量。因为骰子的结果只能是1、2、3、4、5、6这几个可能的取值之一，不存在介于这些值之间的其他可能结果。

另一个例子是抛硬币的结果，它也是一个离散型随机变量，因为硬币的结果只能是正面（Head）或反面（Tail），不存在其他可能性。

在离散型随机变量的情况下，可以通过概率质量函数（Probability Mass Function，PMF）来描述各个可能取值的概率。这个函数指定了每个可能取值发生的概率。

````python
# 计算多个温度数组的平均值
california = np.array([52, 77, 96])
kentucky = np.array([71, 75, 79])

# 打印结果
print(f"Mean California temperature is {california.mean()}")
print(f"Mean Kentucky temperature is {california.mean()}")
````

从均值的角度来讲，这两个数组的均值都是75华氏度。但是这两个数组并非是相等的，这一点查看他们的range()就知道了。

````python
# 对离散进行可视化
plt.plot(range(3), california, color = 'b', label = 'California')
plt.scatter(range(3), california, color = 'b')
plt.plot(range(3), kentucky, color = 'r', linestyle = '-.', label = 'Kentucky')
plt.scatter(range(3), kentucky, color = 'r')
plt.axhline(75, color = 'k', linestyle = '--', label = 'Mean')
plt.legend()
plt.show()
````

很明显的可以看出，红色线与平均线基本持平，但是蓝色线相差甚远。

````python
def sum_of_squares(data):
    mean = np.mean(data)
    return sum(squared_distance(value, mean) for value in data)

california_sum_squares = sum_of_squares(california)
print(f"California's sum of squares is {california_sum_squares}")
````

````python
kentucky_sum_squares = sum_of_squares(kentucky)
print(f"kentucky_sum_squares  is {kentucky_sum_squares }")
````

我们可以发现两者之间的离散程度有着将近30倍的差异，但是这种衡量并不完美。

我们可可以做这样的尝试：

````python
california_duplicated = np.array(california.tolist() * 2)
duplicated_sum_squares = sum_of_squares(california_duplicated)
print(f'Duplicated California sum of squares is {duplicated_sum_squares}')
assert duplicated_sum_squares == 2 * california_sum_squares
````

我们可以查到一个非常离谱的现象：当同样的数据反复出现的时候，平方和会越来越大，如果以平方和作为衡量数据的离散程度的指标，那么这个数据集的离散程度是在扩大的，但是事实并不相符，因为反复出现的数据集只是单个数据集的重复而已，数据的大小并没有发生任何的变化。

为此，需要采取的措施是：将平方和除以数据长度。

````python
value1 = california_sum_squares / california.size
value2 = duplicated_sum_squares / california_duplicated.size

assert value1 == value2
````

这样求得的结果就是方差。

从概念上来讲：方差是每个样本与均值之差的平方值的平均数。

````python
def variance(data):
    mean = np.mean(data)
    return np.mean([squared_distance(value, mean) for value in data])

assert variance(california) == california_sum_squares / california.size
````

可以发现两者的水平是相同的，所以两者的离散程度是相同的。

````python
assert variance(california) == variance(california_duplicated)
````

而california和kentucky数组的方差仍然保持了相同：

````python
california_variance = variance(california)
kentucky_variance = variance(kentucky)
print(f'Kentucky Variance is {kentucky_variance}')
print(f'California Variance is {california_variance}')
````

方差是一个很好的用来观察离散型的度量，可以直接使用np.var函数来进行计算。

````python
assert california_variance == california.var()
assert california_variance == np.var(california)
````

可以发现两者其实是相同的。

方差取决于均值，如果计算的是加权平均值，那么就需要计算加权方差。

````python
def weighted_variance(data, weights):
    mean = np.average(data, weights = weights)
    squared_distances = [squared_distance(value, mean) for value in data]
    return np.average(squared_distances, weights = weights)
````

````python
assert weighted_variance([75, 77], [9, 1]) == np.var(9 * [75] + [77])
````

**正态分布**

正态分布(Normal Distribution)是概率论和统计学中一种重要的概率分布。它描述了一种随机现象的概率分布模式,该模式呈现钟形对称曲线。正态分布具有以下主要特征:

1. 钟形对称曲线：正态分布的概率密度函数曲线呈现对称的钟形,在均值处达到最大值,两侧渐次下降。

2. 均值和标准差：正态分布完全由两个参数决定:均值(μ)和标准差(σ)。均值代表分布的中心位置,标准差描述了分布的离散程度。

3. 68-95-99.7法则：在正态分布中,约68.27%的数据值落在μ±σ的范围内,约95%的数据值落在μ±2σ的范围内,约99.73%的数据值落在μ±3σ的范围内。

4. 广泛应用：正态分布在自然界和社会现象中广泛存在,如身高、智力分数、误差、随机噪声等都近似服从正态分布。因此,它在许多领域都有重要应用,如质量控制、抽样调查、假设检验等。

正态分布的数学表达式为:

$f(x) = (1/σ√(2π)) * e^(-(x-μ)^2/(2σ^2))$

其中,$f(x)$是概率密度函数,x是随机变量的值。

总之,正态分布是一种常见且重要的概率分布模型,广泛应用于自然科学、社会科学和工程技术等领域。

在正态分布中，均值($\mu$)是分布的中心位置，通常也是分布的峰值所在的位置。正态分布是对称的，这意味着其均值、中位数和众数在同一位置，即在其对称轴上。均值的计算是根据分布的定义直接给出的，而不是从峰值"计算"出来的。但对于实际的数据集或者一个已知的概率分布函数，均值的计算方法依赖于分布的性质。

对于一个理论上的连续正态分布，均值 $\mu$ 是分布的参数之一，通常用来定义整个分布，另一个参数是标准差 $\sigma$。这个分布的概率密度函数(PDF)是：

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}
$$

其中，$x$ 是变量，$\mu$ 是均值，$\sigma$ 是标准差。在这个公式中，$\mu$ 直接决定了分布的中心位置，也就是其峰值所在的位置。

如果你有一组数据并且想从这组数据中估计正态分布的均值，你会计算所有数据点的平均值。这个平均值可以看作是分布的峰值所在的估计位置。计算平均值的公式是：

$$
\text{均值} = \mu = \frac{1}{N}\sum_{i=1}^{N}x_i
$$

这里，$N$ 是数据点的总数，$x_i$ 是每个数据点的值。

对于从分布的峰值计算均值的直接问题，实际上在正态分布中，这两个概念是重合的：分布的峰值就在均值所在的位置。而对于其他类型的分布，特别是那些不对称的分布，峰值(众数)和均值可能不会重合，此时均值的计算需要依据分布的具体形式和定义。的计算需要依据分布的具体形式和定义。

2. `var_normal = weighted_variance(bin_edges[:-1], likelihoods)`
   - 这一行计算了直方图数据的加权方差,作为正态分布的方差的估计。

   - 这一行使用f-string格式化字符串,打印出估计的正态分布均值,保留两位小数。

总的来说,这段代码使用直方图的箱边界`bin_edges`和相应的概率密度`likelihoods`,来估计拟合数据的正态分布的均值和标准差。这是通过计算加权平均值和加权方据需要打印出标准差或进一步使用这些估计值。

````python
# 从峰值坐标计算均值和标准差
import math 
peak_x_value = bin_edges[likelihoods.argmax()]
print(f"Mean is approximately {peak_x_value}")
peak_y_value = likelihoods.max()
std_from_peak = (peak_y_value * (2 * math.pi) ** 0.5) ** -1
print(f"Standard deviation is approximately {std_from_peak:.3f}")
````

从峰值的x坐标来估计均值的方法是可行的，因为对于近似正态分布的数据，峰值的x坐标确实大致对应于数据的均值。这就是为什么通过找到likelihoods.argmax()对应的bin_edges值来估计均值的原因，这个方法在某些情况下是有效的。

但是应该注意的是，这个方法假定你已经准确知道了分布是严格的正态分布，并且你能准确地从数据中获得最大概率密度值 $f(μ)$。

或者可以简单的调用 stats.norm.fit(sample_meas)来计算均值和标准差：

````python
fitted_mean, fitted_std = stats.norm.fit(sample_means)
print(f'Mean is approximately {fitted_mean:.2f}')
print(f'Standard deviation is approximately {fitted_std:.3f}')
````

计算的均值和标准差用于重现正态曲线，可以通过调用 stats.norm.pdf(bin_edges, fitted_mean, fitted_std)重新生成曲线。

````python
normal_likelihoods = stats.norm.pdf(bin_edges, fitted_mean, fitted_std)
plt.plot(bin_edges, normal_likelihoods, color='k', linestyle='--',
label='Normal Curve')
plt.hist(sample_means, bins='auto', alpha=0.2, color='r', density=True)
plt.legend()
plt.xlabel('Sample Mean')
plt.ylabel('Relative Likelihood')
plt.show()
````

“相对似然”一词在统计学中有特定含义，尤其在最大似然估计（Maximum Likelihood Estimation, MLE）中。似然函数给出了在给定参数值下观察到数据样本的概率。在许多情况下，似然函数不直接等同于PDF或PMF，因为它是参数的函数，而不是观察数据的函数。相对似然是指某个参数值的似然与最大似然（即似然函数的最大值）的比值。相对似然并不直接等于概率分布，

相对似然可以大于1的原因在于它是一个比率，而不是一个概率。在统计学中，尤其是在进行参数估计或模型比较时，似然函数表示给定某个参数值时，观察到当前数据的"可能性"或"似然度"。当我们谈论相对似然时，我们通常是在比较不同参数值下似然函数的值。

**定义和计算**

相对似然的计算涉及到将某个参数值的似然与最大似然(即似然函数的最大值)进行比较。如果我们有一个参数 $\theta$，并且 $\hat{\theta}$ 是该参数的最大似然估计(MLE)，那么对于任意 $\theta$ 的似然 $L(\theta)$ 与最大似然 $L(\hat{\theta})$ 的比，即相对似然，定义为：

$$
\text{Relative Likelihood}(\theta) = \frac{L(\theta)}{L(\hat{\theta})}
$$

**为什么可以大于1**

在最大似然估计中，我们寻找使似然函数最大化的参数值 $\hat{\theta}$。这个最大似然值 $L(\hat{\theta})$ 作为分母用来标准化其他参数值的似然值。因此，按照这个定义，$\hat{\theta}$ 本身的相对似然是1，因为 $L(\hat{\theta}) / L(\hat{\theta}) = 1$。

对于其他参数值 $\theta$，其相对似然是它们的似然 $L(\theta)$ 与最大似然 $L(\hat{\theta})$ 的比值。然而，由于相对似然是用来比较似然函数值的，它实际上并不是一个概率，所以它的值并不受0到1的概率规则限制。在实际应用中，相对似然通常是小于或等于1的，因为它是与最大似然值(即最"可能"的参数估计)相比较的。如果相对似然大于1，那可能是由于计算或概念上的误解。

总结来说，相对似然之所以通常不会大于1，是因为它是相对于最大似然值本身来衡量的，而最大似然值的相对似然被定义为1。如果你在分析中遇到相对似然大于1的情况，可能需要重新检查计算方法或理解上的假设。

概率密度函数(Probability Density Function, PDF)和概率质量函数(Probability Mass Function, PMF)是用来描述随机变量分布特性的两种函数，它们适用于不同类型的随机变量:连续随机变量和离散随机变量。

**概率密度函数(PDF)**

- **适用对象**：连续随机变量。

- **定义**：对于连续随机变量 $X$，其概率密度函数 $f(x)$ 描述了随机变量取在任一具体值附近的概率密度。PDF本身的值不是概率，因为在连续的情况下，随机变量取任一具体值的概率都是0。相反，PDF在某个区间内的积分表示随机变量落在这个区间内的概率。

- **特性**：
    - $f(x) \geq 0$，对于所有的 $x$。
    - $\int_{-\infty}^{\infty} f(x)dx = 1$，即整个概率空间下，PDF的积分(面积)等于1。

**概率质量函数(PMF)**

- **适用对象**：离散随机变量。

- **定义**：对于离散随机变量 $X$，其概率质量函数 $p(x)$ 描述了随机变量等于任一具体值的概率。PMF直接给出某个值的概率，适用于离散情况。

- **特性**：
    - $0 \leq p(x) \leq 1$，对于所有的 $x$。
    - $\sum_{x} p(x) = 1$，即所有可能值的概率之和等于1。

简而言之，PDF用于连续随机变量，通过积分来找到随机变量落在某个区间的概率；PMF用于离散随机变量，直接给出随机变量取特定值的概率。两者都是描述随机变量如何分布的工具，但因应不同的随机变量类型而有所区别。

````python
# 往右平移0.1
adjusted_likelihoods = stats.norm.pdf(bin_edges, fitted_mean + 0.01,
fitted_std / 2)
plt.plot(bin_edges, adjusted_likelihoods, color='k', linestyle='--')
plt.hist(sample_means, bins='auto', alpha=0.2, color='r', density=True)
plt.xlabel('Sample Mean')
plt.ylabel('Relative Likelihood')
plt.show()
````

**比较两个采样的正态曲线**

SciPy允许我们根据输入的参数探索和调整正态分布的形状。此外，这些输入参数的值取决于如何对随机数据进行采样。

````python
np.random.seed(0)

new_sample_size = 40000
new_head_counts = np.random.binomial(new_sample_size, 0.5, 100000)
new_mean, new_std = stats.norm.fit(new_head_counts / new_sample_size)
new_likelihoods = stats.norm.pdf(bin_edges, new_mean, new_std)

plt.plot(bin_edges, normal_likelihoods, color='k', linestyle='--', 
         label='A: Sample Size 10K')
plt.plot(bin_edges, new_likelihoods, color='b',
         label='B: Sample Size 40K')
plt.legend()
plt.xlabel('Sample Mean')
plt.ylabel('Relative Likelihood')
plt.show()
````

两个正态分布是使用硬币翻转数据生成的。分布 A 是使用每个样本 10,000 次硬币翻转得到的。分布 B 是使用每个样本 40,000 次硬币翻转得到的。两个分布都以均值 0.5 为中心。然而,分布 B 在其中心周围分散得更集中,并且分布 B 的峰值是分布 A 峰值的两倍。根据峰值高度和方差之间的关系,我们可以推断出分布 B 的方差是分布 A 方差的四分之一。


两个正态分布都以样本平均值0.5为中心，然是，具有较大样本的分布图像更窄的聚集在了其峰值周围，随着样本量的增加，峰值的位置保持不变，但是峰值周围的区域宽度在不断的收缩。

峰值变窄导致了质信区间的范围下降，置信区间表示涵盖在正面向上的概率的可能值范围。

开始计算95%的置信区间：

````python
mean, std = new_mean, new_std
start, end = stats.norm.interval(0.95, mean, std)

print(f"The true mean of the sampled binomial distribution is between{start:.3f} and {end:.3f}")
````

我们有95%的把握确信采样的伯努利分布均值介于0.495和0.505之间。事实上，这个平均值正好等于0.5.

````python
assert stats.binom.mean(1, 0.5) == 0.5
````

可以通过比较峰值估计方差的标准差偏移，分布B的峰值是分布A的峰值的两倍，峰值高度与标准差成反比，因此分布B的标准差是分布A的标准差的一半。

因为标准差是方差的平方根，搜易可以推断分布B的方差是分布A的方差的1/4。

因此：将样本量增加刀4倍（从10 000 增加到 40 000）会导致方差减少为1/4。

````python
variance_ratio = (new_std ** 2 ) / (fitted_std ** 2)
print('The ratio of variances is approximately {variance_ratio:.2f}')
````

既然方差与样本量成反比，那么将样本量减少到原先的1/4，就可以将方差增加4倍。

````python
np.random.seed(0)
reduced_sample_size = 2500
head_counts = np.random.binomial(reduced_sample_size, 0.5, 100000)
_, std = stats.norm.fit(head_counts / reduced_sample_size)
variance_ratio = (std ** 2) / (fitted_std ** 2)
print(f"The ratio of variances is approximately {variance_ratio:.1f}")
````

结果与推测的相符合。

但是需要注意样本大小为1的情况：

````python
estimated_variance = (fitted_std ** 2) * 10000
print(f"Estimated variance for a sample size of 1 is {estimated_variance:.2f}")
````

这是因为我们估计的样本大小是1的方差为0.25。但是，如果样本大小为1，那么sample_means数组只是一个随机记录的1和0的序列。

预测样本大小为1的预测方差：

````python
assert stats.binom.var(1, 0.5) == 0.25
````

`stats.binom.var(n, p)` 是用于计算二项分布方差的函数,其中 `n` 是试验的次数, `p` 是每次试验成功的概率。

对于二项分布,方差的公式为:

```
var = n * p * (1 - p)
```

在你给出的例子中, `stats.binom.var(1, 0.5)` 计算的是 `n=1`, `p=0.5` 时的方差。代入公式:

```
var = 1 * 0.5 * (1 - 0.5)
    = 0.5 * 0.5
    = 0.25
```

这个结果符合二项分布方差公式。

换句话说,对于单次伯努利试验(n=1),当成功率p=0.5时,其方差为0.25。这很容易理解,因为当p=0.5时,0和1的概率相等,方差最大。

所以 `stats.binom.var(1, 0.5) == 0.25` 这个结果是正确的,符合二项分布方差的定义。

**计算正态分布曲线下放的面积**

每个梯形的面积等于两个连续似然的总和除以2。


每个梯形的面积等于两个连续似然的总和除以2,解释如下:

在绘制直方图时,整个x轴范围被划分为多个相邻的小区间(bins)。每个bin的宽度表示该区间的范围。而bin的高度对应该区间内数据的计数或密度。因此,bin下的面积代表了该区间内数据的总体贡献。

具体来说,如果我们用$f(x)$表示数据分布的概率密度函数(PDF),那么第$i$个bin的面积可以用如下公式计算:

$$\text{面积} = \frac{f(x_i) + f(x_{i+1})}{2} \times \text{bin宽度}$$

其中:
- $x_i$和$x_{i+1}$分别是第$i$个bin的左右边界
- $f(x_i)$和$f(x_{i+1})$分别是PDF在bin边界处的值
- bin宽度是指$x_{i+1} - x_i$的长度

这个公式本质上是在用一个梯形的面积来近似bin下方的面积。梯形的上底是$f(x_i)$,下底是$f(x_{i+1})$,高是bin宽度。

之所以使用梯形形状,是因为对于连续函数来说,PDF的值在相邻的两点之间是平滑变化的。因此,梯形面积比简单的矩形面积能更好地近似真实情况。

总的来说,这句话解释了如何计算直方图条形下面积的方法,即通过相邻两个PDF值的平均值乘以bin宽度,从而用梯形面积来近似该区间内的总体贡献。梯形面积来近似该区间内的总体贡献。

````python
total_area = np.sum([normal_likelihoods[i: i + 2].sum() / 2
for i in range(normal_likelihoods.size - 1)])

assert total_area == np.trapz(normal_likelihoods)
print(f"Estimated area under the curve is {total_area}")
````

# 统计假设检测

## 引言

**一个具体的例子**

在美国的司法体系中,陪审员们经常不得不做出艰难的抉择。他们需要在审理过程中,根据呈现的证据,在两个相互竞争的假设之间做出选择:

1. 被告无罪
2. 被告有罪

这两个假设的权重并不相等:被告在未被定罪前,都被presumed为无罪。因此,陪审员们初始时假设无罪假设为真。他们只有在检察官的证据确凿的情况下,才能否决无罪假设。但是,证据很少100%确凿,总会存在一些对被告有罪的合理怀疑。这种怀疑会被纳入法律程序中考虑。

陪审团会被指示,如果对被告有罪存在"合理怀疑",就应该接受无罪假设。他们只有在被告"无疑有罪"的情况下,才能否决无罪假设。

"合理怀疑"是一个很抽象的概念,难以精确定义。尽管如此,我们仍然可以分辨出在现实场景中,哪些是合理怀疑,哪些是不合理怀疑。考虑以下两个审判案例:

1. DNA证据直接将被告与罪行联系在一起。但有10亿分之一的几率,DNA不属于被告。
2. 血型证据直接将被告与罪行联系在一起。但有15分之一的几率,血液不属于被告。

在第一种情况下,陪审团不能100%确定被告有罪。有10亿分之一的几率,一个无辜的被告正在受审。然而,这种情况发生的可能性极低,不合理地假设这是真的。因此,陪审团应该否决无罪假设。

而在第二种情况下,合理怀疑的程度要高得多:15分之一的人拥有与被告相同的血型。合理地假设另一个人可能出现在犯罪现场。虽然陪审员可能怀疑被告的无辜,但也会合理地怀疑被告的有罪。因此,除非有更多有罪证据,否则陪审员无法否决无罪假设。

在这两种情况下,陪审员实际上是在执行统计假设检验。这种检验允许统计学家根据不确定的数据,在两个相互竞争的假设之间做出选择。根据测量出的怀疑程度,接受或否决其中一个假设。在本节中,我们将探讨几种著名的统计假设检验技术,从最简单的测试样本均值是否显著偏离既有总体均值开始。

**什么是统计假设检验**

统计假设检验是一种统计推断的方法，用于决定收集到的数据是否提供足够的证据来支持或反对某个特定的假设。它是通过对数据进行分析来测试关于总体参数的假设，如均值、方差等。统计假设检验的目的是帮助研究者判断观察到的数据模式是偶然出现的，还是代表了总体的真实状态。
统计假设检验是一种统计推断的方法,用于决定收集到的数据是否提供足够的证据来支持或反对某个特定的假设。它是通过对数据进行分析来测试关于总体参数的假设,如均值、方差等。统计假设检验的目的是帮助研究者判断观察到的数据模式是偶然出现的,还是代表了总体的真实状态。实验设计、数据质量、结果的实际意义和其他相关因素。


**基本概念**

$H_0$: 零假设通常表示为没有效应、没有差异或状态不变的假设。在假设检验中,零假设是默认假设,假定两个比较的群体之间没有显著差异或者一个治疗效果与无效相同。

$H_a$ 或 $H_1$: 备择假设与零假设相对,通常是研究者希望证明的假设。它表明了存在效应、有差异或状态有变化。

**检验过程:**

1. 提出假设:明确零假设和备择假设。

2. 选择适当的检验统计量:根据数据的类型和研究设计选择一个统计方法(例如,t检验、卡方检验)。

3. 确定显著性水平($\alpha$):显著性水平通常设定为0.05,意味着5%的概率拒绝一个真实的零假设(即犯第一类错误的概率)。

4. 计算检验统计量和P值:通过分析数据计算出检验统计量,并基于此计算P值,P值表示在零假设为真的前提下,观察到的数据(或更极端情况)出现的概率。

5. 做出决策:如果P值小于或等于显著性水平,拒绝零假设,表示有统计学上的显著差异或效应;如果P值大于显著性水平,不拒绝零假设。

**类型错误:**

- 第一类错误:错误地拒绝了真实的零假设(假阳性)。
- 第二类错误:错误地未能拒绝一个假的零假设(假阴性)。

统计假设检验是科学研究中的一项重要工具,它帮助研究者基于数据作出推断和决策。然而,做出决策时也需要考虑实验设计、数据质量、结果的实际意义和其他相关因素。

## 评估样本均值和总体均值之间的差异

假设北达科塔州的五年级学生都在同一天参加了一个同样的评估考试，并且成绩被录入和进行了评估：

````python
population_mean = 80
population_variance = 100
````

现在，我们发现，南达科塔州的一个平均考试成绩为84分的5年纪的班级，该班级有18名学生，成绩比成绩总体高出了4%。

问题：南达科塔州的五年级学生是否接受了更好的教育？还是这只是一个统计上的偶然？

这两个假设实际上是相互竞争的，第一种假设是：南达科塔州和北达科塔州的五年级的学生的成绩并没有什么总体上的区别，即两者的总体均值和总体方差并没有什么统计上的差异。这就是 $H_o$假设，反过来就是$H_1$假设。

假设原假设为真，并且两个达科塔州具有相同的总量均值和方差，那么：

````python
mean = population_mean
sem = (population_variance / 18) ** 0.5
````

sem是均值标准差。

均值标准差（通常称为标准误差）是一个统计学概念，用于描述样本均值的精确度或其对总体均值估计的可靠性。具体来说，它量化了从同一总体中抽取的不同样本的均值之间的变异性。这里有几个关键点来理解均值标准差的用途和重要性：

1. **衡量估计的精确度**：标准误差提供了一种衡量样本均值作为总体均值估计准确度的方法。标准误差越小，意味着样本均值越有可能接近总体均值，从而估计越精确。

2. **构建置信区间**：在统计学中，标准误差用于构建关于总体均值的置信区间。例如，一个95%的置信区间表明，如果我们从总体中重复抽取样本并计算相应的置信区间，那么大约95%的这些区间将包含总体均值。标准误差是这些置信区间宽度的关键决定因素。

3. **进行假设检验**：在进行假设检验时，标准误差用于计算统计量，如t值，进而判断观察到的样本结果是否与某个特定假设（通常是零假设）不一致。如果计算出的统计量超出了某个临界值，我们可能会拒绝零假设，认为样本提供了足够的证据表明与零假设预期的情况有显著差异。

简而言之，均值标准差（标准误差）是一种衡量样本统计量（特别是均值）可靠性的工具，它在估计总体参数、构建置信区间以及进行假设检验中起着核心作用。通过这个指标，研究者可以更好地理解他们的样本结果在统计上的意义和限制。

如果原假设为真，则遇到平均考试成绩至少为84分的概率是： stats.norm.sf(84, mean, sem)

````python
prob_high_grade = stats.norm.sf(84, mean, sem)
print(f"Probability of an average grade >= 84 is {prob_high_grade}")
````

在这个上下文中，`stats.norm.sf()` 函数是从`scipy.stats`库中使用的，用于计算给定随机变量值在正态分布中的生存函数（survival function）值。生存函数也被称为余累积分布函数，它表示随机变量取值大于（或等于）某个特定值的概率。具体到这个例子中，它用于计算平均成绩大于或等于84的概率。

函数的使用方式如下：
```python
stats.norm.sf(x, loc, scale)
```
- `x` 是我们感兴趣的值，在这个例子中是84。
- `loc` 是分布的均值，在这个例子中使用的是`mean`，它表示样本或总体的平均成绩。
- `scale` 是分布的标准差，在正态分布的上下文中，当考虑均值的标准误差（SEM）时，这里使用的是`sem`。

具体到这段代码：
```python
prob_high_grade = stats.norm.sf(84, mean, sem)
```
它计算了在一个正态分布中，平均成绩大于或等于84的概率。这里假设成绩分布符合正态分布，`mean`是这个分布的均值，而`sem`（标准误差）则是其标准差。`stats.norm.sf()`返回的值`prob_high_grade`表示随机从该分布中抽取一个样本，其值大于或等于84的概率。

简而言之，这段代码用于评估在给定均值和标准误差定义的正态分布中，得到平均成绩至少为84的概率。这对于统计推断和假设检验来说是一个非常有用的计算，特别是在评估特定成绩或结果出现的可能性时。

当我们讨论一个样本均值（比如说18个学生的平均成绩）和一个预定值（比如84分）之间的概率关系时，我们实际上是在评估这个样本均值作为对总体均值的估计。在这种情况下，我们感兴趣的是样本均值的抽样分布——即如果我们重复多次抽样，所有这些样本均值将形成的分布。

样本均值的抽样分布具有比单个样本更小的变异性，这个变异性通过标准误差来衡量。因此，当我们利用正态分布来估计样本均值达到或超过某个特定阈值的概率时，应当使用标准误差而不是标准差，因为这反映了样本均值的抽样分布的真实宽度。

简而言之，使用标准误差而不是标准差，是因为在统计分析中，我们通常对样本均值如何代表总体均值感兴趣，而标准误差正是衡量这种代表性精确度的指标。通过将标准误差应用于正态分布函数，我们能够更准确地评估样本均值达到或超过特定值的概率。







````python
# 双边检测，查看是否具有负面效应
prob_low_grade = stats.norm.cdf(76, mean, sem)
print(f"Probability of an average grade <= 76 is {prob_low_grade}")
````

从结果来看，双边是相同的。这种相同源于正态曲线的对称形状。累计分布函数和生存函数反映在平均值上的镜像。

````python
for x in range(-100, 100):
    sf_value = stats.norm.sf(mean + x, mean, sem)

assert sf_value == stats.norm.cdf(mean - x, mean, sem)

plt.axvline(mean, color='k', label='Mean', linestyle=':')

x_values = range(60, 101)

plt.plot(x_values, stats.norm.cdf(x_values, mean, sem), label='Cumulative Distribution')
plt.plot(x_values, stats.norm.sf(x_values, mean, sem), label='Survival Function', linestyle='--', color='r')

plt.xlabel('Sample Mean')
plt.ylabel('Probability')
plt.legend()

plt.show()
````

生存函数是统计学中的一个概念，用于描述在随机事件中，事件发生的时间超过某个特定时间点的概率。更具体地说，它是在可靠性理论、生存分析和各种时间至事件分析中使用的，以衡量个体（例如，患者、机械部件等）"生存"（即未发生事件）的概率随时间的变化。生存函数通常用 \\(S(t)\\) 表示，其中 \\(t\\) 是时间。

**定义**

给定随机变量 \\(T\\)，表示事件发生的时间，生存函数 \\(S(t)\\) 定义为在时间 \\(t\\) 或之后事件尚未发生的概率。数学上，它可以表达为：

\\[ S(t) = P(T > t) \\]

这里，\\(P(T > t)\\) 表示时间 \\(T\\) 大于某个特定值 \\(t\\) 的概率。

**与其他函数的关系**

生存函数与累积分布函数（CDF, \\(F(t)\\)）密切相关，后者表示随机变量取值小于或等于某个值的概率。生存函数和累积分布函数之间的关系可以表示为：

\\[ S(t) = 1 - F(t) \\]

此外，生存分析中另一个重要的函数是风险函数（hazard function），它描述了给定生存到时间 \\(t\\) 后，在下一个很短时间间隔内事件发生的条件概率密度。

**应用**

生存函数在医学、工程、社会科学、金融等多个领域都有应用。例如，在医学研究中，生存函数可以用来估计在接受特定治疗后患者存活超过一定时间的概率；在工程中，可以用来估计设备或系统在特定时间内不发生故障的概率。

总的来说，生存函数是一个强大的工具，用于分析和理解在一定时间范围内未发生特定事件的概率。通过生存函数的分析，研究人员和决策者可以更好地理解和预测事件的发生，从而制定更有效的策略和决策。

累积分布函数（CDF）和生存函数在本质上是互补的，因为它们分别描述了随机变量取值在某个阈值之下和之上的概率。这种互补关系导致了它们在图形上呈现出镜像的特性。具体来说：

- **累积分布函数（CDF）** \\(F(t)\\) 表示随机变量 \\(T\\)（代表时间或任何其他度量）取值小于或等于 \\(t\\) 的概率。数学上，它定义为 \\(F(t) = P(T \\leq t)\\)。

- **生存函数** \\(S(t)\\) 表示随机变量 \\(T\\) 取值大于 \\(t\\) 的概率。它可以表示为 \\(S(t) = P(T > t)\\)。

由于 \\(T\\) 取值小于或等于 \\(t\\)（累积到 \\(t\\) 的概率）和 \\(T\\) 取值大于 \\(t\\) 的概率在逻辑上是互补的（它们加起来等于1），因此有 \\(S(t) = 1 - F(t)\\)。这表明如果你知道了 \\(T\\) 的CDF，你也自然知道了它的生存函数，反之亦然。

**镜像的含义**

当我们说CDF和生存函数是镜像的，我们是指它们的关系可以通过概率的完全性（所有概率加起来等于1）来直观理解。CDF随 \\(t\\) 增加而单调增加，从0增加到1；而生存函数随 \\(t\\) 增加而单调减少，从1减少到0。这意味着，如果你在同一个坐标系中绘制这两个函数，CDF从左下角向右上角增长，而生存函数从左上角向右下角减少，它们在 \\(y\\) 轴上的值互补，形成一种"镜像"效果。

**应用意义**

这种镜像关系不仅在数学上很优雅，而且在实际应用中非常有用。通过分析生存函数，研究人员可以直观地看到在某个时间点之后事件尚未发生的概率，这在生存分析、可靠性工程和风险管理中是非常重要的。同样，CDF提供了在给定时间点或之前事件已经发生的累积概率，这对于概率分析和统计推断来说同样重要。理解这两个函数之间的关系有助于更全面地分析和解释随机事件的行为。

prob_high_grade和prob_low_grade两者进行求和，得到的结果就是2*prob_high_grade，从概念上来讲，总和表示当原假设为真的时候观察到与总体均值存在极端差异的概率。统计学家将这种原假设驱动的概率成为p值，也就是p-value。

````python
p_value = prob_low_grade + prob_high_grade
assert p_value == 2 * prob_high_grade == 2 * prob_low_grade
print(f"The p-value is {p_value}")
````

也就是说，大约有9%的机会可以随机观察到极端成绩，因此，我们没有足够的证据能够拒绝H0假设。

假设分数不是85而是84，代码为：

````python
def compute_p_value(observed_mean, population_mean, sem):
    mean_diff = abs(population_mean - observed_mean)
    prob_high = stats.norm.sf(population_mean + mean_diff, population_mean, sem)
    return 2 * prob_high

new_p_value = compute_p_value(85, mean, sem)
print(f"The updated p-value is {new_p_value}")
````

现在可能性只有3%了，所以我们可以考虑拒绝原假设，即H0假设，因此，我们可以认为教学方和影响成绩具有着统计上的显著性。

如果设置一个非常严苛的p值，即将p值设置为0.001，那么：

````python
for grade in range(80, 100):
    p_value = compute_p_value(grade, mean, sem)
    if p_value < 0.01:
        break
print(f'An average grade of {grade} leads to a p-value of {p_value}')
````

如果使用新的p-value，那么至少需要88分（如果满分为100分）的平均成绩才能够拒绝原假设。

选择不同的显著性水平（如0.01和0.05）作为判断拒绝零假设（\(H_0\)）的标准，反映了在统计检验中容忍犯第一类错误（假阳性，即错误地拒绝了真实的零假设）的概率的不同。这个选择直接影响了研究的严格性和结论的可靠性。

**显著性水平0.05**

- **更宽松的标准**：选择0.05作为显著性水平意味着愿意接受5%的概率犯第一类错误。这是社会科学、生物学和许多其他领域中的常见做法，因为它提供了平衡研究灵敏度（检测实际效应的能力）和控制错误的需要之间的妥协。
- **更高的灵敏度**：在某些情况下，使用0.05的显著性水平可以增加研究发现统计显著效应的可能性，但同时也增加了发生假阳性结果的风险。

**显著性水平0.01**

- **更严格的标准**：选择0.01作为显著性水平意味着只有当观察到的数据（或更极端的数据）在零假设为真的情况下出现的概率小于1%时，才会拒绝零假设。这反映了对犯第一类错误的更低容忍度，通常在需要更高证据标准的研究中采用，如医学试验。
- **更低的灵敏度**：虽然使用0.01的显著性水平可以减少假阳性的风险，但它也可能导致研究对于检测真实存在的效应变得不够敏感（即增加了犯第二类错误的风险，或假阴性，即错误地接受了一个假的零假设）。

**选择的影响**

- **研究领域的差异**：不同的研究领域和具体的研究情境可能要求不同的显著性水平。例如，在探索性研究中可能接受更高的犯错误概率，而在某些需要高度可靠性的研究（如药物测试）中，则可能需要更低的显著性水平。
- **权衡和决策**：选择显著性水平涉及到灵敏度与特异性之间的权衡。研究者需要根据研究的目的、可能的后果以及犯错误的容忍度来做出决策。

总之，选择不同的\(p\)-值阈值（如0.01与0.05）反映了在统计检验中对发现真实效应的敏感性与控制错误之间的不同平衡点。更严格的显著性水平可以减少假阳性的风险，但也可能错过一些真实的发现。

下面是一个总结第一类错误和第二类错误的表格，包括它们的定义、示例以及如何控制这些错误的概率：

| 错误类型 | 定义 | 示例 | 控制方法 |
|---------|------|------|----------|
| 第一类错误（假阳性） | 错误地拒绝了真实的零假设 $H_0$ | 在药物试验中，判断一个无效的药物为有效 | 通过设定较低的显著性水平 $\alpha$（如0.05或0.01）来降低犯此错误的概率 |
| 第二类错误（假阴性） | 错误地接受了一个假的零假设 $H_0$ | 在药物试验中，判断一个有效的药物为无效 | 通过增加样本大小或提高检验的统计功效（1 - $\beta$ )来降低犯此错误的概率 |

## 数据捕捞：过度采样将会导致错误的结论

**两个关于P-value的解释**

断言一：P值表示的是原假设为真的概率

**P值表示的是原假设为真的概率**：这个断言是**错误的**。$p-value$不是原假设为真的概率。$p-value$实际上衡量的是在原假设（零假设 $H_0$为真的条件下，观察到的数据或更极端数据出现的概率。它不直接告诉我们原假设为真或假的概率，而是告诉我们在假定原假设为真的情况下，观察到当前结果或更极端结果的可能性有多大。

断言二：P值表示的是原假设成立的情况之下观察到偏差的概率

**P值表示的是原假设成立的情况之下观察到偏差的概率：**这个断言是正确的。正确理解p-值的含义是关键的：它是在原假设（或零假设）为真的条件下，通过实验或观察得到当前统计结果或更极端结果的概率。如果这个概率很小（比如小于研究者事先设定的显著性水平$α$，通常是0.05或0.01），我们就有理由拒绝原假设，因为在其为真的假定下获得这样的结果是不太可能的。

**重点：通过反复计算最终会观察到一个较低的p值**


这句话涉及到一个在统计分析和实验设计中重要但有时被忽视的概念，即如果一个实验不断重复进行，最终可能会因为随机变异而观察到一个显著的（即较低的）\\(p\\)-值，即使实际上没有真正的效应存在。这里的关键在于理解\\(p\\)-值的随机性和重复实验对统计推断的影响。

- **\\(p\\)-值的随机性**：\\(p\\)-值并不是一个固定的属性，而是基于样本数据的一个随机变量。因此，即使在相同的实验条件下，不同的样本可能会产生不同的\\(p\\)-值。

- **重复实验的影响**：如果我们对相同的假设进行多次独立测试，每次都计算\\(p\\)-值，那么由于随机抽样误差，某些测试可能会产生低\\(p\\)-值，即使总体上没有实际效应存在。这意味着，通过足够多次的尝试，我们几乎可以保证至少有一次测试会得到显著的结果。

- **显著性水平**：通常，研究者可能会设置\\(p\\)-值的显著性阈值（比如0.05），意味着他们愿意接受5%的概率犯第一类错误（错误地拒绝真实的零假设）。如果一个实验反复进行足够多次，那么仅仅由于随机变异，我们可以期待大约每20次实验中就有一次会得到\\(p < 0.05\\)，即使实际上没有效应。

- **误解的风险**：这种现象揭示了一个统计学上的警示：仅仅因为一个实验结果显著（\\(p\\)-值低）并不一定意味着有一个真实的效应或关联。这也是为什么科学研究强调结果的复制重要性——即不同研究者在独立的研究中重复发现相同的结果——以确保发现不仅仅是随机波动的结果。

因此，"通过反复计算最终会观察到一个较低的\\(p\\)-值"这句话提醒我们，对单一实验结果的解释需要谨慎，尤其是在没有复制研究支持的情况下。它也强调了多次测试后进行校正（如Bonferroni校正）的重要性，以避免因多重比较而增加的第一类错误率。重比较而增加的第一类错误率。

准确地说，如果$p$-值的阈值设置为0.05，并且进行足够多的实验，其中零假设（$H_0$）实际上是真的（即没有真实的效应存在），那么预期大约有5%的实验会观察到$p$-值小于或等于0.05。这意味着这些实验结果会被错误地视为具有统计显著性，从而导致第一类错误，即错误地拒绝了真实的零假设。

重要的是要理解这里的5%并不是指观察到$p$-值**恰好**为0.05的概率，而是观察到$p$-值**小于或等于**0.05的概率。这个阈值以下的任何$p$-值都被视为足够的证据来拒绝零假设，假定实验是在零假设实际为真的条件下进行的。

因此，当你进行大量实验时，即使不存在真实效应，你也可以期待大约5%的实验会得到显著的结果，仅仅由于随机变异。这正是为什么在实际研究中，单一的实验结果需要通过独立复制来验证，以确保发现不是偶然结果的重要原因之一。因之一。

**使用Bonferroni校证**

Bonferroni校正是一种统计方法，用于减少在多重比较中犯第一类错误（假阳性）的风险。当在同一个数据集上进行多次统计测试时，即使没有任何实际的效应存在，也更有可能得到统计上显著的结果。Bonferroni校正通过调整显著性水平来应对这个问题，以保持整体的第一类错误率在一个可接受的范围内。


假设你有n个独立的假设测试，而你希望整体的第一类错误率（即至少有一个假阳性的概率）不超过α（通常是0.05）。Bonferroni校正简单地将每个单独测试的显著性水平设置为$α/n$。这种方法基于Bonferroni不等式，是保守的，意味着它倾向于减少假阳性的风险，但可能增加假阴性（第二类错误）的风险。

优点：Bonferroni校正简单、直观，易于应用。它非常有效地控制了家族错误率（family-wise error rate, FWER），即一个或多个类型I错误在多个假设检验中发生的概率。

缺点：由于其保守性，Bonferroni校正可能导致功效降低，特别是当进行大量测试时，这可能使得发现真实效应变得更加困难。

````python
# 使用 Bonferroni 校正调整显著性
num_planned_experiments = 49
significance_level = 0.05 / num_planned_experiments
````

## 有回放的自举法：当总体方差未知时检验假设

假设有个水族箱：

````python
fish_lengths = np.array([46.7, 17.1, 2.0, 19.2, 7.9, 15.0, 43.4,
                         8.8, 47.8, 19.5, 2.9, 53.0, 23.5, 118.5,
                         3.8, 2.9, 53.9, 23.9, 2.0, 28.2])

assert fish_lengths.mean() == 27
````

现在得到了可靠的消息：热带鱼平均长度为37厘米。

我们并不能够直接凭感觉，因为这不符合统计学原则；但是我们不知道总体方差，而样本太少所以并不能够直接认为样本方差是总体方差。

所以需要采用的方法：bootstrapping with replacement。（有回放的自举法）

````python
np.random.seed(0)
random_fish_length = np.random.choice(fish_lengths, size=1)[0]
sampled_fish_lengths = [random_fish_length]
````

随机抽取一条鱼，测量记录，然后放回去，然后再随机抽取伊条鱼（就算抽到了之前放回去的那条也无所谓）。

````python
# 重复随机采样20条鱼的长度
np.random.seed(0)
for _ in range(20):
    random_fish_length = np.random.choice(fish_lengths, size=1)[0]
    sampled_fish_lengths.append(random_fish_length)
````

因为是随机采样，所以sampled_fish_lengths和fish_lengths的元素并不完全相同，因此平均值也不尽相同。

````python
sample_mean = np.mean(sampled_fish_lengths)
print(f"Mean of sampled fish lengths is {sample_mean:.2f} cm")
````

可以发现存在着一点儿差异。

````python
# 使用NumPy进行放回采样
np.random.seed(0)
new_sampled_fish_lengths = np.random.choice(fish_lengths, size=20, replace=True)  # size = 20 不是一次性抽取20个，而是总共20次
new_sample_mean = new_sampled_fish_lengths.mean()
print(f"Mean of the new sampled fish lengths is {new_sample_mean:.2f} cm")
````

接着重复15000次来得到结果：

````python
np.random.seed(0)

sample_means = [np.random.choice(fish_lengths, size=20, replace=True).mean()
                for _ in range(150000)]

likelihoods, bin_edges, _ = plt.hist(sample_means, bins='auto', edgecolor='black', density=True)

plt.xlabel('Binned Sample Mean')
plt.ylabel('Relative Likelihood')
plt.show()
````

````python
abs(stats.skew(sample_means))
````

如果绝对值大于了0.04（还是0.4？）的偏态数据，那么这个偏态数据不太可能来自于正态分布的数据。

非对称直方图不能够用来进行正态建模，但是直方图代表了一个连续的概率分布，鱼所有连续分布一样，直方图可以映射到概率密度函数，累计分布函数和生存函数。

````python
random_variable = stats.rv_histogram((likelihoods, bin_edges))
plt.plot(bin_edges, random_variable.pdf(bin_edges))
plt.hist(sample_means, bins='auto', alpha=0.1, color='r', density=True)
plt.xlabel('Sample Mean')
plt.ylabel('Relative Likelihood')
plt.show()
````

````python
rv_mean = random_variable.mean()

print(f"Mean of the distribution is approximately {rv_mean:.2f} cm")
plt.axvline(random_variable.mean(), color='k', label='Mean', linestyle=':')
plt.plot(bin_edges, random_variable.cdf(bin_edges), label='Cumulative Distribution')

plt.plot(bin_edges, random_variable.sf(bin_edges), label='Survival', linestyle='--', color='r')
plt.xlabel('Sample Mean')
plt.ylabel('Probability')
plt.legend()

plt.show()
````

不能够简单的将生存函数的输出翻倍来计算p值，并且非对称当中，生存函数本身不足以计算尾部概率。

但是幸运的是：可以通识使用生存函数和累计分布函数来揭示计算观察的概率。然后通过这些概率，来进行显著性的评估。

使用random_variables.sf(37)来获得观察到的样本的长度至少为37厘米的概率；同理使用random_variables.sf(17)来获得样本均值小于或者等于17的概率。

````python
prob_extreme= random_variable.sf(37) + random_variable.cdf(17)
print("Probability of observing an extreme sample mean is approximately "
      f"{prob_extreme:.2f}")
````

因此，样本平均值（27）和总体平均值（37）之间的差异没有统计学意义。

prob_extreme是p值的一个伪装，当原假设为真的时候，在10%的采样中，样本均值与总体均值之间的差异将至少为10厘米。这个p值0.1大于阈值0.05，因此不能够拒绝原假设。

````python
# 使用自举法估计SEM
estimated_sem = random_variable.std()
p_value = compute_p_value(27, 37, estimated_sem)
print(f"P-value computed from estimated SEM is approximately {p_value:.2f}")
````

如同预期的一样，计算出的p值大约为0.1。


第一种方法:

1. 从数据中有放回地抽样,重复数万次,获得一系列样本均值。

2. 根据这些样本均值生成一个直方图。

3. 使用stats.rv_histogram方法将直方图转换为分布。

4. 使用生存函数和累积分布函数计算分布曲线的左右两个极端下方的面积。

第二种方法看起来稍微简单一些:

1. 从数据中有放回地抽样,重复数万次,获得一系列样本均值。

2. 计算均值的标准差来近似估计标准误差(SEM)。

3. 使用估计得到的SEM,通过我们的compute_p_value函数进行基本的假设检验。

接着采用第三种方法，这种方法不需要使用直方图，也不依赖自定义的计算值函数：

````python
# 使用大树定理，样本量足够大，则观察到的事件频率近似于事件发生的概率
number_extreme_value = 0
for sample_mean in sample_means:
    if not 17 < sample_mean < 37:
        number_extreme_value += 1

p_value = number_extreme_value / len(sample_means)
print(f"P-value is approximately {p_value:.2f}")
````

最后，使用random_variable对象进行假设检验。

````python
# 将极值定义为至少19的差异
p_value = random_variable.sf(19) + random_variable.cdf(-19)
print(f"P-value is approximately {p_value:.2f}")
````

P值约为0.04，低于显著性阈值0.05，因此可以拒绝H0假设，也就是这两个水族馆的鱼不来自于同一个总体。

````python
number_extreme_values = 0.0
for min_diff in mean_diffs:
    if not -19 < min_diff < 19:
        number_extreme_values += 1

p_value = number_extreme_values / len(mean_diffs)
print(f"P-value is approximately {p_value:.2f}")
````

## 根据均值差异计算p值

首先将蓝色删除，留下其他的29种颜色，然后对表进行转置，从而通过列名访问颜色对应的数据。

````python
df_not_blue = df.T.drop(columns='Blue')
print(df_not_blue.head(2))
````

将这些百分比与蓝色百分比进行比较，更准确的来讲，想知道是否有平均点击率与蓝色的平均点击率在统计上存在着不同的颜色。

````python
def permutation_test(data_array_a, data_array_b):
    data_mean_a = data_array_a.mean()
    data_mean_b = data_array_b.mean()
    extreme_mean_diff = abs(data_mean_a - data_mean_b)
    total_data = np.hstack([data_array_a, data_array_b])
    number_extreme_values = 0.0
    for _ in range(30000):
        np.random.shuffle(total_data)
        sample_a = total_data[:data_array_a.size]
        sample_b = total_data[data_array_a.size:]
        if abs(sample_a.mean() - sample_b.mean()) >= extreme_mean_diff:
            number_extreme_values += 1

    p_value = number_extreme_values / 30000
    return p_value
````

我们没有总体均值，因此，最好的选择是进行置换检验。

接着对颜色进行置换检验：

````python
np.random.seed(0)

blue_clicks = df.T.Blue.values
color_to_p_value = {}

for color, color_clicks in df_not_blue.items():
    p_value = permutation_test(blue_clicks, color_clicks)
    color_to_p_value[color] = p_value

sorted_colors, sorted_p_values = zip(*sorted(color_to_p_value.items(),
                                             key=lambda x: x[1]))

plt.figure(figsize=(3, 10))
sns.heatmap([[p_value] for p_value in sorted_p_values],
            cmap='YlGnBu', annot=True, xticklabels=['p-value'],
            yticklabels=sorted_colors)

plt.show()
````

这段代码执行了以下步骤，用于比较`df`中蓝色点击数（`blue_clicks`）与其他颜色点击数的差异，并通过置换测试(permutation test)计算每个颜色与蓝色之间的显著性差异（即p值），最后将结果通过热图展示出来。具体步骤如下：

1. **设置随机数种子**：`np.random.seed(0)` 确保了接下来的随机操作，如置换测试中的数据洗牌，可以被重复。这使得结果在每次运行时都是一样的。

2. **获取蓝色点击数**：`blue_clicks = df.T.Blue.values` 从DataFrame `df` 中提取蓝色点击数。这里首先对`df`进行转置（`.T`），然后选择蓝色（`Blue`）列，并获取其值（`.values`）。

3. **初始化一个空字典**：`color_to_p_value = {}` 用于存储每个颜色与蓝色点击数比较的p值。

4. **遍历非蓝色点击数**：`for color, color_clicks in df_not_blue.items():` 遍历一个包含非蓝色点击数的DataFrame或字典`df_not_blue`。对于每种颜色（`color`）和其点击数（`color_clicks`），执行以下步骤：

    - 调用`permutation_test`函数，将`blue_clicks`（蓝色点击数）和`color_clicks`（当前颜色的点击数）作为输入，计算它们之间的p值。
    
    - 将计算得到的p值与对应颜色存入`color_to_p_value`字典。

5. **排序和解包**：`sorted_colors, sorted_p_values = zip(*sorted(color_to_p_value.items(), key=lambda x: x[1]))` 根据p值将颜色排序，并将排序后的颜色和p值分别解包到两个变量中。这里使用了`sorted`函数和一个lambda函数来指定按照字典的值（p值）进行排序。

6. **绘制热图**：接下来的代码块利用`matplotlib.pyplot`和`seaborn`库创建一个热图，展示每种颜色与蓝色点击数之间显著性差异的p值：

    - `plt.figure(figsize=(3, 10))` 创建一个大小为3x10英寸的图形。
    
    - 使用`seaborn.heatmap`函数创建热图，其中`[[p_value] for p_value in sorted_p_values]`是一个二维列表，表示热图的数据矩阵。`cmap='YlGnBu'`定义了使用的颜色映射，`annot=True`表示在热图中注释数据值。`xticklabels=['p-value']`和`yticklabels=sorted_colors`分别设置了x轴和y轴的标签。

    - `plt.show()`显示绘制的热图。

通过这个过程，代码有效地比较了蓝色点击数与其他颜色点击数的统计学显著性，并以视觉化的方式展现了每种颜色与蓝色点击数差异的显著性大小。这可以帮助识别哪些颜色的点击数与蓝色有显著性差异，可能对进一步的数据分析和决策制定很有帮助。


热力图告诉我们以下信息：

- y轴上列出的颜色代表与参考组（本例中为蓝色）比较的不同组别。
- 每个颜色名称旁边的p值代表该颜色的点击数据与蓝色点击数据之间差异的统计显著性。
- p值从非常小的数（表示非常显著的差异）到接近1的数（表示不显著的差异）不等。
- 热力图通过颜色深浅来表示p值的大小：深蓝色表示p值很小，显著性高；浅蓝色到白色表示p值较大，显著性低。
- 通常在统计测试中，p值小于0.05（或在某些领域小于0.01）被认为是统计学上显著的。在这个图中，较暗的色块（如黑色、石灰色、粉色等）的p值远小于0.05，表示与蓝色的点击数差异非常显著。而接近图底部的颜色（如青色、蓝绿色）的p值接近1，表示与蓝色的差异不显著。
- 结果排序是按照p值的大小进行的，从最小到最大。

这张热力图可以帮助我们理解哪些颜色的点击数与蓝色存在显著性差异，对于进行进一步的数据分析和决策制定可能非常有帮助。

如果我们前面讲的，只要我们做了足够多的实验，我们总是能够遇到p值低的情况，所以我们需要使用Bonferroni校正。

````python
significance_level = 0.05 / 29
print(f"Adjusted significance level is {significance_level}")
if color_to_p_value['Ultramarine'] <= significance_level:
    print("Our p-value is statistically significant")
else:
    print("Our p-value is not statistically significant")
````

得到的p值指出在统计学上并不显著，因为Fred做了太多次实验，因此无法得到有意义的结论。

````python
# 尝试将蓝色与其他5中深浅不一的蓝色进行比较
hypothetical_sig_level = 0.05 / 5
print(f"Hypothetical significance level is {hypothetical_sig_level}")
if color_to_p_value['Ultramarine'] <= hypothetical_sig_level:
    print("Our hypothetical p-value would have been statistically significant")
else:
    print("Our hypothetical p-value would not have been statistically significant")
````

在假设条件下，我们的结果是具有统计意义的。

遗憾的是，我们不能利用这些假设条件来降低显著性水平。我们无法保证重复实验能够再次得到0.0034的p值。p值会波动，过多的实验会增加得到不可靠波动的风险。考虑到弗雷德已经做了很多实验，我们简直无法得出具有统计意义的结论。

然而，情况并非全无希望。紫蓝色仍然是蓝色的有希望的替代品。那么，弗雷德是否应该进行替换呢？也许。让我们考虑两个备选情景。在第一种情况下，零假设成立，即蓝色和紫蓝色具有相同的人群平均值。在这种情况下，将紫蓝色替换为蓝色不会影响广告点击率。在第二种情况下，紫蓝色更高的点击率实际上是具有统计意义的。如果是这样，那么替换成紫蓝色将会带来更多的广告点击。

因此，弗雷德通过将所有广告设置为紫蓝色，将一无所失，而有一切可得。从逻辑上讲，弗雷德绝对应该将蓝色换成紫蓝色。但如果他进行替换，一些不确定性仍将存在；弗雷德永远不会知道紫蓝色是否真的比蓝色带来更多点击。如果弗雷德的好奇心占了上风，他唯一的选择是进行另一个实验。在这个实验中，一半显示的广告将是蓝色，另一半将是紫蓝色。


弗雷德的软件会展示这些广告同时记录所有的点击和浏览。然后我们可以重新计算p值，并将其与0.05的适当显著性水平进行比较。因为只进行了一次实验，所以不需要进行Bonferroni校正。在p值比较之后，弗雷德最终将知道紫蓝色是否比蓝色表现更好。
