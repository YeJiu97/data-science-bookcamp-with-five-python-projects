# 对数据进行聚类

**关于聚类**

聚类是一种无监督学习算法，它的主要目的是将相似的数据点分组到一起。在聚类过程中，算法会根据数据点之间的相似性将它们分配到不同的群组或“簇”中，数据点之间的相似性通常是通过一种距离度量来计算的，如欧几里得距离或曼哈顿距离。聚类算法尝试使同一簇内的数据点尽可能相似，而不同簇的数据点则尽可能不同。

聚类在许多领域都有广泛应用，包括但不限于：

1. **市场细分**：通过识别具有相似需求和偏好的客户群体，企业可以更有效地定位市场和定制产品。
2. **社交网络分析**：在社交网络分析中，聚类可以帮助识别具有相似兴趣或互动模式的用户群体。
3. **图像分割**：在图像处理领域，聚类算法可以用来识别图像中的相似像素，进而实现图像分割。
4. **基因表达数据分析**：在生物信息学中，聚类可以用于分析和归类具有相似表达模式的基因。

常见的聚类算法包括K-means聚类、层次聚类、DBSCAN等。每种算法都有其特点和适用场景，选择哪种算法通常取决于数据的特性和分析的目标。

## 使用中心性进行聚类

直观来讲：每个组中的所有数据点都应该围绕着一个中心值进行聚集。

同时：两个不同组的测量值应该聚集在两个不同的均值周围。

因此：我们可以使用两个不同的中心性来区分两个不同的组。

我们可以使用丢飞镖来进行模拟：

````python
# 导入numpy作为必备库
import numpy as np

# 设置一个随机种子，用来进行之后的复现
np.random.seed(0)

# 将原点设置为平均值
mean = 0

# 方差的值设置为2
variance = 2

# np.random.normal()函数可以用来生成符合高斯分布/正态分布的随机样本
x = np.random.normal(mean, variance ** 0.5)  # 标准差
y = np.random.normal(mean, variance ** 0.5)

print(f"The x coordinate of a randomly thrown dart is {x:.2f}")
print(f"The y coordinate of a randomly thrown dart is {y:.2f}")
````

注：实际上可以使用 np.random.multivariate_normal方法来更加有效的模拟飞镖的位置，因为这个方法能够从多元正态分布中选择单个随机点。

````python
import matplotlib.pyplot as plt
np.random.seed(1)

# 这里采用二位坐标，而不再是一维坐标系
bulls_eye1 = [0, 0]
bulls_eye2 = [6, 0]

# 两者共同的结果
bulls_eyes = [bulls_eye1, bulls_eye2]

# 用来表示飞镖的x和y坐标
x_coordinates, y_coordinates = [], []

# 开始遍历
for bulls_eye in bulls_eyes:
    # 遍历5000次
    for _ in range(5000):
        # 投掷一次飞镖，生成符合正态分布的x和y的值
        x = np.random.normal(bulls_eye[0], variance ** 0.5)
        y = np.random.normal(bulls_eye[1], variance ** 0.5)
        # 将其加入到坐标中
        x_coordinates.append(x)
        y_coordinates.append(y)
# 注意：我们不要将每个飞镖手投掷的飞镖单独的存储到一个列表中，因为我们将会将两者的投掷结果绘制到同一个图像中
plt.scatter(x_coordinates, y_coordinates)
plt.show()
````

途中出现了两组有部分重叠的飞镖结果，这两组结果代表10 000次飞镖投掷实验。

一半的飞镖瞄准左侧的靶心，另一半的飞镖瞄准右侧的靶心。

让我们将每个飞镖分配到最近的靶心，首先定义一个nearest_bulls_eye函数，它将一个包含飞镖x和y坐标的dart列表作为输入。

该函数会返回飞镖更加接近与哪个靶心，我们用欧式距离测量飞镖的距离，欧式距离是两点直接的标准直线距离。

**欧式距离**

欧氏距离（Euclidean distance）是数学中的一种距离定义，用于度量多维空间中两个点之间的直线距离。它是最直观、最常用的一种距离度量方式，在许多领域如机器学习、数据分析、地理空间分析等都有广泛应用。

给定两个点 $P = (p_1, p_2, ..., p_n)$ 和 $Q = (q_1, q_2, ..., q_n)$ 在一个 $n$维空间中，欧氏距离 $d(P, Q)$可以通过下面的公式计算：

$d(P, Q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2 + ... + (p_n - q_n)^2}$

换句话说，欧氏距离是由勾股定理（Pythagorean theorem）推广而来，用于计算两点之间的直线距离。在二维空间中，这个概念等同于两点之间的直线距离；而在三维空间中，它是空间中两点间的实际距离。

在机器学习和数据科学中，欧氏距离常用于计算样本点之间的相似度，例如在聚类算法（如 K-means）和分类算法中。简单来说，如果两个点之间的欧氏距离较小，则可以认为它们在几何上较为接近，反之亦然。

````python
# 将飞镖归到最近的靶心
from scipy.spatial.distance import euclidean

def nearest_bulls_eye(dart):
    # 使用从SciPy导入的欧几里得函数获得飞镖和每个靶心之间的欧几里得距离
    distance = [euclidean(dart, bulls_e) for bulls_e in bulls_eyes]  # 列表推导式，bulls_eyes在之前已经定义了
    # np.argmin()会给出最小值，这里就是最短距离
    return np.argmin(distance)  # 返回这个结果

darts = [[0, 1], [6, 1]]  # 两个飞镖

for dart in darts:
    index = nearest_bulls_eye(dart)
    print(f'The dart at position {dart} is closest to bulls-eye {index}')
````

接着我们将这个函数应用到所有的飞镖坐标，然后使用两种颜色来进行绘图：

````python
# 定义这个函数
def color_by_cluster(darts):
    # 使用之前的函数对于darts中的dart进行遍历和选择最近的中心，使用欧几里得距离
    nearest_bulls_eyes = [nearest_bulls_eye(dart) for dart in darts]
    # 开始进行遍历，如果最近的中心等于之前设置的中心，那么将其归属到这个中心的簇当中
    for bs_index in range(len(bulls_eyes)):
        selected_darts = [darts[i] for i in range(len(darts))
                          if bs_index == nearest_bulls_eyes[i]]
        # 这里进行转置，可以用来分离每个飞镖的x和y坐标
        x_coordinates, y_coordinates = np.array(selected_darts).T
        # 将其绘制为散点图
        plt.scatter(x_coordinates, y_coordinates,
                    color=['g', 'k'][bs_index])
    # 绘制最终的图像
    plt.show()

darts = [[x_coordinates[i], y_coordinates[i]]
         for i in range(len(x_coordinates))]

color_by_cluster(darts)
````

1. 使用一个之前定义的 nearest_bulls_eye 函数，该函数对于给定的一个飞镖位置（dart），计算并返回它最近的靶心索引。对于输入的 darts 列表中的每一个元素（即每个飞镖的位置），都调用 nearest_bulls_eye 函数，生成一个新的列表 nearest_bulls_eyes，其中包含每个飞镖最近的靶心索引。
2. 代码接着遍历所有的靶心（由全局变量 bulls_eyes 表示）。对于每一个靶心，找出所有最近的飞镖（这是通过比较 nearest_bulls_eyes 列表中的索引和当前靶心的索引实现的）。然后，提取这些飞镖的 x 坐标和 y 坐标，并使用 plt.scatter 方法绘制这些点。每个靶心对应的飞镖组使用不同的颜色（在示例中使用了绿色和黑色，通过列表 ['g', 'k'] 索引选择颜色）。
3. 最后，使用 plt.show() 方法显示图形，可视化地展示出每个飞镖靠近的靶心和它们的分组情况。

```python
        x_coordinates, y_coordinates = np.array(selected_darts).T
```

这行代码使用了 Python 的多维数组操作和元组解包来同时分离出 `x` 和 `y` 坐标。这里的 `selected_darts` 是一个二维列表（或者 NumPy 数组），其中每个元素都是表示一个飞镖位置的 `[x, y]` 列表（或数组）。具体步骤如下：

1. **转换为 NumPy 数组**：首先，`np.array(selected_darts)` 将 `selected_darts` 列表（或列表的列表）转换成一个 NumPy 数组。假设 `selected_darts` 包含了 `n` 个 `[x, y]` 对，那么转换后的数组将是一个 `n x 2` 的二维数组，其中 `n` 是飞镖的数量。

2. **转置数组**：接着，使用 `.T` 属性（表示矩阵的转置）来转置这个二维数组。在转置之后，原来的列变成了行：第一行现在包含所有的 `x` 坐标，第二行包含所有的 `y` 坐标。

3. **元组解包**：最后，`x_coordinates, y_coordinates = ...` 这部分代码通过元组解包将转置后的数组的第一行赋值给 `x_coordinates`，第二行赋值给 `y_coordinates`。因此，`x_coordinates` 成为一个包含所有 `x` 坐标的数组，而 `y_coordinates` 成为一个包含所有 `y` 坐标的数组。

这种方法非常高效地分离出了 `x` 和 `y` 坐标，因为它利用了 NumPy 库的矩阵操作能力，避免了显式的循环遍历每个坐标对。

我们利用颜色将飞镖的坐标均匀的分成两个聚类。

但是这是我们知道坐标的情况之下，如果不知道呢？

````python
# 一个方法是随机选取一个点进行聚类
bulls_eyes = np.array(darts[:2])  # 这里选择了darts的前两个点来作为靶心
color_by_cluster(darts)
````

通常来讲，随机选取靶心得到的结果并不是非常的理想。

一个纠正的方法是：计算所有右侧飞镖的平均坐标，然后将这个平均坐标作为靶心坐标。

````python
# 由平均值来计算靶心来分配飞镖
def update_bulls_eyes(darts):
    updated_bulls_eyes = []
    nearest_bulls_eyes = [nearest_bulls_eye(dart) for dart in darts]
    for bs_index in range(len(bulls_eyes)):
        selected_darts = [darts[i] for i in range(len(darts))
                         if bs_index == nearest_bulls_eyes[i]]
        x_coordinates, y_coordinates = np.array(selected_darts).T
        # 根据对于特定靶心的所有飞镖坐标计算x和y的平均值，然后将这个平均值作为新的靶心位置
        mean_center = [np.mean(x_coordinates), np.mean(y_coordinates)]
        updated_bulls_eyes.append(mean_center)

    return updated_bulls_eyes

bulls_eyes = update_bulls_eyes(darts)
color_by_cluster(darts)
````

上述代码片段是一个名为 `update_bulls_eyes` 的函数定义，旨在更新一系列靶心的位置。这个过程通过考虑一组飞镖的落点（即 `darts`），来计算每个靶心新的坐标位置。函数开始时，首先通过对 `darts` 中的每个点调用 `nearest_bulls_eye` 函数来确定每个点最近的靶心。虽然具体的 `nearest_bulls_eye` 函数实现没有给出，但我们可以推断它返回与给定飞镖点最近的靶心的索引。

接下来，函数进入一个循环，遍历每个靶心的索引。对于每个靶心，函数筛选出所有最近靶心为当前靶心的飞镖点。这是通过检查前面计算的最近靶心索引列表来完成的，只有当飞镖点的最近靶心索引与当前靶心索引相匹配时，该飞镖点才会被选中。

然后，对于每个靶心，函数计算所有选中飞镖点的 x 坐标和 y 坐标的平均值。这个平均坐标点（即均值中心）成为更新后的靶心位置。这样，通过计算属于同一靶心的所有飞镖点的均值中心，我们可以得到一个新的靶心位置，更好地代表了该靶心对应的飞镖点的集中趋势。

最终，这个过程产生了一个更新后的靶心位置列表，即 `updated_bulls_eyes`，它包含了根据当前飞镖点的分布计算出的每个靶心的新位置。这个更新过程是迭代优化算法（如 K-means 聚类）的核心步骤之一，用于根据数据点的实际分布逐步调整簇中心，直到达到最佳的聚类效果。

代码的最后两行实际上使用了 `update_bulls_eyes` 函数来更新靶心位置，并尝试用 `color_by_cluster` 函数给 `darts` 着色，后者的目的可能是为了可视化每个飞镖点根据最近的靶心归属于不同的簇或组。这种可视化有助于理解聚类算法如何将数据点分组，并观察靶心位置的更新如何影响这些分组。

从图像来看，大体上和之前完美的分类近似了，但是还是存在着一些差异的，这里可以使用迭代来达到完美聚类的结果，进行10次迭代：

````python
for i in range(10):
    bulls_eyes = update_bulls_eyes(darts)

color_by_cluster(darts)
````

## K-means：一种将数据分组为K个中心组的聚类算法

**什么是K-means聚类**

K-means 是一种广泛使用的聚类算法，旨在将数据点划分为 K 个簇，其中 K 是用户指定的簇的数量。算法通过迭代优化来最小化每个点到其簇中心的距离平方和，从而寻找数据点的自然分组。K-means 算法的基本步骤如下：

1. **初始化**：随机选择 K 个数据点作为初始的簇中心（centroid）。
2. **分配步骤**：将每个数据点分配给最近的簇中心，形成 K 个簇。
3. **更新步骤**：对于每个簇，重新计算簇中所有点的平均位置，将该平均位置设为新的簇中心。
4. **重复步骤 2 和 3**：不断重复分配和更新步骤，直到簇中心不再发生变化或变化非常小，或者达到预设的迭代次数，此时算法收敛。

K-means 算法简单而强大，但它也有一些局限性：
- 需要事先指定 K 值，而在实际应用中，最优的 K 值往往不是那么容易确定。
- 对于非球形的数据分布，K-means 的效果可能不是很好。
- 对于不同大小或不同密度的簇，K-means 也可能表现不佳。
- 初始簇中心的选择可能会对算法的最终结果产生较大影响。虽然有一些方法（如 K-means++ 初始化方法）旨在通过更加精心的初始簇中心选择来改善这一点，但这个问题并没有完全解决。

尽管存在这些局限性，K-means 由于其实现简单和效率高，在许多实际应用中仍然非常受欢迎，包括市场细分、文档聚类、图像分割等领域。

**使用scikit-learn来进行K-means聚类**

````python
from sklearn.cluster import KMeans

cluster_model = KMeans(n_clusters=2)
assigned_bulls_eyes = cluster_model.fit_predict(darts)

print("Bull's-eye assignments:")
print(assigned_bulls_eyes)
````

最后，打印出每个数据点被分配的簇标签。在这个例子的输出 [0 0 0 ... 1 1 1] 中，省略号表示输出结果在中间部分被缩略。这个序列显示了数据集中的每个点被分配的簇标签。序列开始时连续的 0 表示这些点都被分配到了第一个簇，而序列结束时的连续 1 则表示这些点被分配到了第二个簇。这种分配表明，根据 KMeans 算法计算出的簇中心，数据点被成功划分成了两组。

此结果是 KMeans 聚类算法应用于 darts 数据集的直接输出。算法根据每个点到两个簇中心的距离来决定每个点的簇归属，最终达到将数据集分成两部分的目的。这种分割方式对于理解数据集内在的结构、执行市场细分、图像分割等任务非常有用。





接着开始根据聚类分配为飞镖尽心着色，从而验证结果。

````python
# 绘制k-means聚类分配
for bs_index in range(len(bulls_eyes)):  # bulls_eyes是由0和1组成的
    selected_darts = [darts[i] for i in range(len(darts))  
                      if bs_index == assigned_bulls_eyes[i]]
    # 这里也是进行分离
    x_coordinates, y_coordinates = np.array(selected_darts).T
    # 分配进去
    plt.scatter(x_coordinates, y_coordinates,
                color=['g', 'k'][bs_index])
plt.show()
````

我们使用了聚类模型来定位了数据中的质心，现在可以使用这些质心分析模型中以前没有出现过的新数据点。

````python
new_darts = [[500, 500], [-500, -500]]
new_bulls_eye_assignments = cluster_model.predict(new_darts)

for i, dart in enumerate(new_darts):
    bulls_eye_index = new_bulls_eye_assignments[i]
    print(f"Dart at {dart} is closest to bull's-eye {bulls_eye_index}")
````

**使用肘部法选择最佳K值**

K-means算法依赖于输入的K，当事先不知道数据中真实聚类的数量的时候，这将会成为一个严重的问题。

肘部法（Elbow Method）是一种用于确定最佳聚类数量（即k值）的方法，广泛应用于K-means聚类等算法中。这种方法的基本思想是通过变化聚类数量（k值），计算每个k值对应的成本（通常是簇内误差平方和SSE），然后选择一个“肘点”，即成本下降速率显著变缓的点作为最佳聚类数量。选择k值的具体步骤如下：

1. **选择k的范围**：首先确定k值的范围，比如从1到10。
2. **对每个k值进行聚类**：对于每一个k值，使用聚类算法（如K-means）进行聚类。
3. **计算成本**：计算每次聚类的成本，常用的成本计算方法是簇内误差平方和（SSE），即每个点到其簇中心的距离的平方和。
4. **绘制肘部曲线**：在图表上绘制每个k值对应的成本。随着k值的增加，每个簇中的点会更加紧密，因此成本会下降。但是，在某一点之后，增加k值带来的成本下降会变得非常缓慢。
5. **选择肘点**：找到成本下降突然变缓的点，即“肘点”，这个点就是最佳的聚类数量k。

肘部法是一种直观的方法，但它也有局限性，有时候肘点可能不太明显，或者存在多个潜在的肘点，这时就需要结合业务背景或其他方法（如轮廓系数法）来辅助确定最佳的k值。

````python
# 绘制k值的惯性
k_values = range(1, 10)
inertia_values = [KMeans(k).fit(darts).inertia_
                  for k in k_values]
 
plt.plot(k_values, inertia_values)
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
````

可以发现生成的图像很像是肘部弯曲的手臂，肘部的位置指向了2。

````python
new_bulls_eye = [12, 0]
for _ in range(5000):
    x = np.random.normal(new_bulls_eye[0], variance ** 0.5)
    y = np.random.normal(new_bulls_eye[1], variance ** 0.5)
    darts.append([x, y])
 
inertia_values = [KMeans(k).fit(darts).inertia_
                  for k in k_values]
 
plt.plot(k_values, inertia_values)
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
````

在上述代码中，我们王darts列表中添加了第三个中心，可以发现惯性图也发生了改变，出现了一个新的转折点，其最低倾角指向了K值3。

从本质上来讲，惯性图跟踪了每一个增量K所捕获的离散度，连续K值之间惯性的迅速减少意味着离散的数据点被分配到一个更紧密的聚类中。当惯性曲线变得平缓的时候，惯性的减少逐渐失去其影响，从垂直下降到平缓角度的转变导致了途中肘部形状的出现。

## 使用密度来发现聚类

假设一位天文学家在太阳系的遥远边缘发现了一颗新行星，这颗行星与土星非常的相似，这个行星有着多个环，并且围绕着其中心以恒定轨道旋转。

每个环都是以数以千技的岩石形成的，将这些岩石建模为由x和y坐标定义的单个点。

````python
# 这里使用 scikit-learn的make_circles函数生成由许多岩石组成的3个岩环
from sklearn.datasets import make_circles

# 坐标
x_coordinates = []
y_coordinates = []

# 开始生成
for factor in (.3, .6, 0.99):  # 遍历三个不同的 factor 值（0.3、0.6、和0.99）
    rock_ring, _ = make_circles(n_samples = 800,  # # n_samples=800 指定每个环有800个样本点
                                factor = factor,  # factor 参数控制内外圈的间隔大小，这里的三个值分别生成了三个不同间隔的岩环
                               noise = 0.03,  # noise=0.03 添加了一点噪声，使得岩石的位置有轻微的随机偏差，更加自然
                                random_state= 1)  # 确保每次生成的数据都是一样的，以便于重现结果
    for rock in rock_ring:
        x_coordinates.append(rock[0])
        y_coordinates.append(rock[1])

# 绘制图
plt.scatter(x_coordinates, y_coordinates)
plt.show()
````

从图像来看，可以很明显的发现图中由三个环，所以我们可以通过将K设置为3来使用K-means搜索这桑聚类。

````python
# 使用 K-means 来对环进行聚类
rocks = [[x_coordinates[i], y_coordinates[i]]
        for i in range(len(x_coordinates))]
rock_clusters = KMeans(3).fit_predict(rocks)

colors = [['g', 'k', 'y'][cluster] for cluster in rock_clusters]

plt.scatter(x_coordinates, y_coordinates, color = colors)
plt.show()
````

得到的结果和我们所期望的结果并不相同。K-means将数据分解成3个堆成的段，每个段跨越了多个环，而我们所希望的是每个环被单独的认为是一个段，进而将数据分割出来。

我们需要设计一种算法：在密集的空间区域中对数据进行聚类。

这就要求：我们定义给定区域是密集的还是稀疏的。

密度的定义如下所示：一个点至少与Y个其他点的距离小于X时，这个点才处于密集区域当中。

我们将：
- X称为：epsilon
- Y称为：min_points

这里采用的设置为：
- epsilon = 0.1
- min_points = 10

也就是说：如果一颗岩石与至少10颗其他岩石的距离小于0.1，则这颗岩石处于密集空间当中。

````python
# 指定密度
epsilon = 0.1
min_points = 10
````

我们可以分析 rocks 列表中第一块岩石对应的密度：

````python
rocks[0]
````

首先在rocks[0]位置以epsilon为半径搜索所有其他岩石，这里将这些相邻岩石的索引存储在neighbor_indices列表中。

````python
# 寻找rocks[0]的邻居
neighbor_indices = [i for i, rock in enumerate(rocks[1:])
                   if euclidean(rocks[0], rock) <= epsilon]
````

````python
# 查看一下neineighbor_indices
neighbor_indices[:10]
````

接着将邻居的数量和min_points进行比较，以确定岩石[0]是否位于空间的密集区域。

````python
# 计算rocks[0]的密度
num_neighbors = len(neighbor_indices)
print(f"The rock at index 0 has {num_neighbors} neighbors.")
 
if num_neighbors >= min_points:
    print("It lies in a dense region.")
else:
    print("It does not lie in a dense region.")
````

索引0处的岩石位于空间的密集区域。

rocks[0]的另据是否也存在于那个密集的空间区域？这是一个非常难回答的问题。

因为每个邻居都可能存在少于自己的min_points个邻居的情况，根据严格的密度定义，我们不会将这些邻居视为在密集区域当中。

然而，这将会导致一种荒谬的情况，其中密集区域仅由一个点组成：rocks[0]。

解决方法是：通过更新密度定义避免这种荒谬的结果：

- 如果一个点在以epsilon为半径的范围内由min_points个邻居，那么这个点是在密集区域中的
- 密集区域中的点的每个邻居也被聚类在这个密集区域中

根据更新的定义，可以把rocks[0]和它的邻居组合成一个单一的密集聚类：

````python
# 创建密集聚类
dense_region_indices = [0] + neighbor_indices
dense_region_cluster = [rocks[i] for i in dense_region_indices]
dense_cluster_size = len(dense_region_cluster)
print(f"We found a dense cluster containing {dense_cluster_size} rocks")
````

位于索引0的岩石及其邻居构成了一个包含41个元素的单一密集簇。如果邻居的邻居也属于密集区域，那么根据我们更新后的定义，这些岩石也属于密集簇。因此，通过分析额外的邻近点，我们可以扩大密集区域簇的大小。





````python
dense_region_indices = set(dense_region_indices)

for index in neighbor_indices:
    point = rocks[index]
    neighbors_of_neighbors = [i for i, rock in enumerate(rocks)
                              if euclidean(point, rock) <= epsilon]
    if len(neighbors_of_neighbors) >= min_points:
        dense_region_indices.update(neighbors_of_neighbors)

dense_region_cluster = [rocks[i] for i in dense_region_indices]
dense_cluster_size = len(dense_region_cluster)
print(f"We expanded our cluster to include {dense_cluster_size} rocks")
````

## DBSCAN：一种基于空间密度对数据进行分组的聚类算法

一个简短的视频：https://www.bilibili.com/video/BV114411M7ci/?spm_id_from=333.337.search-card.all.click&vd_source=70cc82c6f851aaa826e5c863112d2113

DBSCAN算法是一种基于密度的聚类方法，可以识别出具有不同形状和大小的簇，并且对噪声和孤立点具有良好的鲁棒性。DBSCAN的主要优势在于它不需要事先指定簇的数量，与K-means这类需要预定义簇数量的算法不同。它主要通过以下两个参数来控制聚类过程：


- ε (epsilon): 一个距离阈值，用于确定点的邻域。如果两个点之间的距离小于或等于ε，则它们被认为是邻居。- 
MinPts (最小点数): 一个点的邻域中最少需要有MinPts个点（包括该点自身），该点才能被认为是核心点。这意味着核心点的ε-邻域内至少有MinPts个点。

DBSCAN算法的基本步骤如下：

1. **从数据列表中随机选择一个点的坐标。**
   
   这是初始步骤，旨在随机挑选一个起始点来开始聚类过程。

2. **获取该点ε距离内的所有邻居。**
   
   通过计算和其他所有点之间的距离，找出在给定的ε（epsilon，一个距离阈值）范围内的所有邻居点。

3. **如果发现的邻居少于min_points，则使用不同的随机点重复步骤1。否则，将点及其邻居归为一个簇。**
   
   这里的min_points是一个阈值，表示一个点及其ε-邻域内至少需要有多少个点（包括自身），这个点才能被考虑为一个核心点，从而开始形成一个簇。如果不满足这个条件，这个点被视为噪声点，算法会重新选择另一个随机点开始新一轮的探索。

4. **在所有新发现的邻居上迭代重复步骤2和3。所有相邻的密集点都合并到簇中。当簇不再扩大时，迭代终止。**
   
   这个步骤说明了聚类的扩展过程：如果一个点被确定为核心点，那么它的所有邻居（满足ε和min_points条件）都被加入到同一个簇中，然后算法会继续检查这些邻居的邻居，以此类推，直到没有新的点可以被添加到簇中。

5. **提取完整簇后，对所有尚未分析密度的数据点重复步骤1-4。**
   
   一旦一个簇被完全提取，算法不会停止，而是继续从剩余的未处理点中选择新的点进行分析，直到所有点要么被归入某个簇，要么被识别为噪声点。

这个过程的关键在于，它能够识别和提取数据中基于密度的簇，而不需要预先知道簇的数量，同时有效地处理噪声数据。这种类型的算法特别适合处理具有复杂结构或噪声较多的数据集。


DBSCAN程序可以使用不到20行代码来进行实现，但是，任何的基本实现在rocks列表中都会运行的很慢。

````python
# 可以使用sklearner自带的函数
# 使用DBSCAN对环进行聚类
from sklearn.cluster import DBSCAN

cluster_model = DBSCAN(eps=epsilon, min_samples=min_points)  # 设置参数

rock_clusters = cluster_model.fit_predict(rocks)  # 开始进行拟合

# 设置颜色
colors = [['g', 'y', 'k'][cluster] for cluster in rock_clusters]

# 绘图
plt.scatter(x_coordinates, y_coordinates, color=colors)
plt.show()
````

下面是一个使用Markdown格式的表格，对DBSCAN算法和K-means算法进行比较：

| 特性         | DBSCAN                                        | K-means                                    |
|--------------|------------------------------------------------|--------------------------------------------|
| 聚类数量     | 不需要预先指定                                | 需要预先指定聚类的数量（k值）               |
| 簇的形状     | 能够识别任何形状的簇                          | 倾向于识别凸形状的簇                       |
| 噪声和孤立点 | 能有效处理噪声和孤立点                        | 对噪声和孤立点敏感                         |
| 算法复杂度   | 较高，因为需要计算和比较所有点之间的距离      | 较低，主要计算每个点到簇中心的距离         |
| 结果的一致性 | 通常产生相同的结果，取决于ε和MinPts的选择     | 可能因初始化的不同而产生不同的结果         |
| 应用场景     | 适合数据集具有复杂结构或含有噪声的情况        | 适合数据集规模较大、簇形状接近于球形的情况 |

这个表格总结了DBSCAN算法和K-means算法在几个关键方面的不同。DBSCAN是基于密度的聚类方法，不需要预设簇的数量，适合于形状复杂、含有噪声的数据集。而K-means算法需要预先定义聚类的数量，适合于簇形状较为规则、数据集较大的情况。每种算法都有其适用的场景，选择哪一种算法取决于具体的数据特性和需求。

注意：与K-means不同，拟合的DBSCAN模型不能够重新应用于全新的数据，想法，需要合并新旧数据并且从头开始进行重新聚类拟合。

原因：K-means中心可以轻松的与其他数据点进行比较，但是，额外的数据点将会影响先前存在的数据的密度分布，这回迫使DBSCAN重新计算所有聚类。

````python
# 使用DBSCAN查找异常值
noisy_data = rocks + [[1000, -1000]]
clusters = DBSCAN(eps = epsilon, 
                  min_samples=min_points).fit_predict(noisy_data)
assert clusters[-1] == -1
````

DBSCAN技术的另外一个优点是：这个技术不依赖均值，但是K-means算法要求我们计算分组点的平均坐标。

如果坐标不满足欧几里得定理，则均值并不是有效的，不应该使用K-means算法来进行计算。

除了欧几里得距离之外，还有其他的距离指标：余弦相似，曼哈顿距离，切比雪夫距离等等。

**基于非欧几里得距离的聚类方法**

曼哈顿距离：https://www.bilibili.com/video/BV1y54y1m7pz/?spm_id_from=333.337.search-card.all.click&vd_source=70cc82c6f851aaa826e5c863112d2113

假设我们正在访问曼哈顿，并希望知道从帝国大厦到哥伦布圆环的步行距离。

帝国大厦位于第34街和第五大道的交叉口。同时，哥伦布圆环位于第57街和第八大道的交叉口。曼哈顿的街道和大道总是相互垂直。

这让我们可以将曼哈顿表示为一个2D坐标系统，其中街道位于x轴上，大道位于y轴上。在这种表示下，帝国大厦位于坐标(34, 5)，哥伦布圆环位于坐标(57, 8)。

我们可以轻松计算两个坐标点之间的直线欧几里得距离。然而，这个最终的长度是无法通过的，因为高耸的钢铁建筑占据了每个城市街区轮廓线内的区域。


一个更正确的解决方案限于穿过形成城市网格的垂直人行道的路径。这样的路线要求我们在第五大道和第三大道之间走3个街区，然后在第34街和第57街之间走23个街区，总共26个街区的距离。

曼哈顿的平均街区长度为0.17英里，所以我们可以估算步行距离为4.42英里。让我们直接使用一个泛化的manhattan_distance函数来计算那个步行距离。

````python
# 计算曼哈顿距离
def manhattan_distance(point_a, point_b):
    num_blocks = np.sum(np.absolute(point_a - point_b))
    return 0.17 * num_blocks

x = np.array([34, 5])
y = np.array([57, 8])

distance = manhattan_distance(x, y)

print(f"Manhattan distance is {distance} miles")
````

现在，假设我们希望对多个曼哈顿位置进行聚类。

我们假设每个簇包含一个点，该点与其他三个簇中的点的步行距离在一英里以内。

这个假设让我们可以使用scikit-learn的DBSCAN类来应用DBSCAN聚类。

在DBSCAN的初始化过程中，我们将eps设置为1，min_samples设置为3。

此外，我们在初始化方法中传入metric=manhattan_distance。metric参数将欧几里得距离替换为我们自定义的距离度量，因此聚类距离正确地反映了城市内基于网格的限制。

下面的代码会聚类曼哈顿坐标，并将它们连同其簇标记一起绘制在网格上。

````python
points = [[35, 5], [33, 6], [37, 4], [40, 7], [45, 5]]

clusters = DBSCAN(eps=1, min_samples=3,
                  metric=manhattan_distance).fit_predict(points)

for i, cluster in enumerate(clusters):
    point = points[i]
    if cluster == -1:
        print(f"Point at index {i} is an outlier")
        plt.scatter(point[0], point[1], marker='x', color='k')
    else:
        print(f"Point at index {i} is in cluster {cluster}")
        plt.scatter(point[0], point[1], color='g')

plt.grid(True, which='both', alpha=0.5)
plt.minorticks_on()
plt.show()
````

前三个位置属于同一个簇，而其余点是离群点。我们能否使用K-means算法检测到这个簇呢？

也许可以。毕竟，我们的曼哈顿街区坐标可以取平均值，使其与K-means实现兼容。

如果我们将曼哈顿距离换成一个不同的度量，平均坐标不那么容易获得呢？

让我们定义一个非线性的距离度量，具有以下属性：
- 如果两点的所有元素都是负的，它们之间的距离为0单位；
- 如果所有元素都是非负的，它们之间的距离为2单位；
- 否则，它们之间的距离为10单位。

给定这种荒谬的距离度量，我们能计算任意两点的平均值吗？我们不能，K-means也无法应用。算法的一个弱点是它依赖于平均距离的存在。不像K-means，DBSCAN算法不要求我们的距离函数是线性可分的。因此，我们可以轻松地使用我们荒谬的距离度量运行DBSCAN聚类。


K-means算法在处理具有线性可分性和可以计算平均值的数据集时效果较好。这是因为K-means的核心是计算簇中点的均值（或质心），然后根据点到这些质心的距离重新分配点到最近的簇。如果距离度量是非线性的，特别是如果它不允许简单地计算点的平均位置，那么K-means算法就不适用。

另一方面，DBSCAN算法基于密度的概念，查找高密度区域并将其归为同一个簇，同时识别低密度区域的点作为离群点。DBSCAN不依赖于计算点的平均位置，因此可以使用各种距离度量，包括非常规和非线性的度量。

````python
def ridiculous_measure(point_a, point_b):
    is_negative_a = np.array(point_a) < 0
    is_negative_b = np.array(point_b) < 0
    if is_negative_a.all() and is_negative_b.all():
        return 0
    elif is_negative_a.any() or is_negative_b.any():
        return 10
    else:
        return 2
 
points = [[-1, -1], [-10, -10], [-1000, -13435], [3,5], [5,-7]]
 
clusters = DBSCAN(eps=.1, min_samples=2,
                  metric=ridiculous_measure).fit_predict(points)
 
for i, cluster in enumerate(clusters):
    point = points[i]
    if cluster == -1:
        print(f"{point} is an outlier")
    else:
        print(f"{point} falls in cluster {cluster}")
````

DBSCAN聚类方法：

- `dbscan_model = DBSCAN(eps=epsilon, min_samples=min_points)` — 创建一个DBSCAN模型来通过密度进行聚类。定义密集点为在距离epsilon内至少有min_points个邻居的点。这些邻居被认为是与该点属于同一个簇的一部分。
- `clusters = dbscan_model.fit_predict(data)` — 使用初始化的DBSCAN对象在输入数据上执行DBSCAN。clusters数组包含簇ID。数据data[i]的簇ID等于clusters[i]。未聚类的离群点被分配一个ID为-1。
- `clusters = DBSCAN(eps=epsilon, min_samples=min_points).fit_predict(data)` — 在一行代码中执行DBSCAN，并返回结果簇。
- `dbscan_model = DBSCAN(eps=epsilon, min_samples=min_points, metric=metric_function)` — 创建一个DBSCAN模型，其中距离度量由自定义的度量函数定义。metric_function距离度量不需要是欧几里得的。

DBSCAN确实有一些缺点。该算法旨在检测具有相似点密度分布的簇。

然而，现实世界中的数据密度是多变的。

例如，曼哈顿的比萨饼店分布比加利福尼亚州橙县的更密集。

因此，在选择能够让我们同时对两个地点的店铺进行聚类的密度参数时，我们可能会遇到困难。

这突出了算法的另一个限制：DBSCAN需要为eps和min_samples参数赋予有意义的值。特别是，不同的eps输入将极大地影响聚类的质量。不幸的是，没有一种可靠的方法可以估计适当的eps。虽然文献中偶尔提到某些启发式方法，但它们的好处很小。

大多数时候，我们必须依靠对问题的直观理解，为这两个DBSCAN参数赋予实际的输入值。例如，如果我们要聚类一组地理位置，我们的eps和min_samples值将取决于这些位置是分布在整个地球上，还是仅限于某个单一的地理区域。在每个实例中，我们对密度和距离的理解都会有所不同。

一般来说，如果我们要聚类分布在地球上的随机城市，我们可以将min_samples和eps参数分别设置为三个城市和250英里。这假设每个簇中的城市与至少其他三个聚类城市之间的距离在250英里以内。对于更局部的位置分布，需要更低的eps值。

## 使用Pandas分析聚类

到目前为止，我们保持数据输入和聚类输出是分开的。

例如，在我们的岩环分析中，输入数据位于rocks列表中，而聚类输出位于rock_clusters数组中。

要追踪坐标和簇，我们需要在输入列表和输出数组之间映射索引。因此，如果我们希望提取簇0中的所有岩石，我们必须获取所有rock_clusters[i] == 0的rocks[i]实例。这种索引分析复杂难懂。通过在单个Pandas表中结合坐标和簇，我们可以更直观地分析聚类的岩石。





````python
import pandas as pd
x_coordinates, y_coordinates = np.array(rocks).T
df = pd.DataFrame({'X': x_coordinates, 'Y': y_coordinates,
                   'Cluster': rock_clusters})
````

````python
df_cluster = df[df.Cluster == 0]
plt.scatter(df_cluster.X, df_cluster.Y)
plt.show()
````

Pandas允许我们获得包含任一单个簇中元素的表格。或者，我们可能想要获取多个表格，每个表格对应一个簇ID。在Pandas中，这是通过调用`df.groupby('Cluster')`来实现的。`groupby`方法将创建三个表格：每个簇一个。它将返回一个可迭代对象，遍历簇ID与表格之间的映射。让我们使用`groupby`方法来迭代我们的三个簇。

````python
for cluster_id, df_cluster in df.groupby('Cluster'):
    if cluster_id == 0:
        print(f"Skipping over cluster {cluster_id}")
        continue
    print(f"Plotting cluster {cluster_id}")
    plt.scatter(df_cluster.X, df_cluster.Y)
plt.show()
````

Pandas groupby 方法让我们迭代地检查不同的集群。 这在我们的案例研究 3 分析中可能很有用。
